```python
#!/usr/bin/env python3
# -*- coding: utf-8 -*-
"""
Enhanced BERT Training with Bias Control and CIRBE Alignment
Author: Yanis Bencheikh et al.
Date: December 2024
"""

# Cell 1: Installation
# Run this first in Colab
"""
!pip install -q torch>=2.0.0
!pip install -q transformers==4.30.0
!pip install -q tokenizers==0.13.3
!pip install -q datasets==2.12.0
!pip install -q scikit-learn matplotlib seaborn tqdm pandas numpy scipy
!pip install -q nltk
!pip install -q requests
"""

# Cell 2: Imports
import os
import json
import time
import hashlib
import random
from datetime import datetime
import numpy as np
import pandas as pd
import torch
import torch.nn as nn
import torch.nn.functional as F
import nltk
import requests
from collections import defaultdict
from torch.utils.data import DataLoader, Dataset, random_split
from torch.cuda.amp import GradScaler, autocast
from pathlib import Path
from typing import Dict, List, Tuple, Optional, Set
import matplotlib.pyplot as plt
import seaborn as sns
from tqdm.auto import tqdm
from sklearn.metrics.pairwise import cosine_similarity
from scipy import stats
import warnings
warnings.filterwarnings('ignore')

import functools

# Cell 3: Setup JSON encoder and utilities
class NumpyEncoder(json.JSONEncoder):
    def default(self, obj):
        if isinstance(obj, np.integer):
            return int(obj)
        elif isinstance(obj, np.floating):
            return float(obj)
        elif isinstance(obj, np.ndarray):
            return obj.tolist()
        elif isinstance(obj, (datetime, pd.Timestamp)):
            return obj.isoformat()
        return super(NumpyEncoder, self).default(obj)

# Monkey patch json.dump to always use NumpyEncoder
original_json_dump = json.dump
json.dump = functools.partial(original_json_dump, cls=NumpyEncoder)

def safe_load_json(filepath):
    """Safely load JSON file with error handling"""
    try:
        with open(filepath, 'r') as f:
            return json.load(f)
    except (json.JSONDecodeError, FileNotFoundError) as e:
        print(f"⚠️ Warning: Could not load {filepath}: {str(e)}")
        print("  Starting with fresh configuration")
        return None

# Download NLTK resources
nltk.download('wordnet', quiet=True)
nltk.download('omw-1.4', quiet=True)

# Transformers imports
from transformers import (
    BertConfig, BertForMaskedLM,
    get_linear_schedule_with_warmup
)
from tokenizers import BertWordPieceTokenizer, Tokenizer
from tokenizers.processors import BertProcessing
from nltk.corpus import wordnet

print("✅ Packages imported successfully!")

# Cell 4: Configuration
class ImprovedTrainingConfig:
    """Configuration for enhanced training with CIRBE alignment"""
    # Architecture
    model_type = "bert"
    vocab_size = 30000
    max_length = 512

    # Training
    batch_size = 32
    learning_rate = 1e-4
    num_epochs = 15
    warmup_steps = 10000
    weight_decay = 0.01
    gradient_accumulation_steps = 4
    max_grad_norm = 1.0

    # Logging and saving
    logging_steps = 100
    save_steps = 1000
    eval_steps = 500

    # MLM
    mlm_probability = 0.15

    # GPU optimizations
    fp16 = True
    gradient_checkpointing = False

    # New parameters
    random_seed = 42
    validation_split = 0.1
    stratified_sampling = True
    temporal_validation = True
    bias_monitoring = True
    bias_check_frequency = 500

    # Early stopping based on bias
    bias_early_stopping = True
    bias_threshold_stop = 0.35

    # Word pairs aligned with CIRBE
    bias_word_pairs = [
        ("systemic", "temporary"),
        ("systemic", "individual"),
        ("insufficient", "adequate"),
        ("insufficient", "generous"),
        ("discriminatory", "standard"),
        ("flagrant", "technical"),
        ("alarming", "anecdotal"),
        ("mandatory", "voluntary"),
        ("fundamental", "cultural"),
        ("violated", "respected"),
        ("limited", "adequate"),
        ("underfunded", "supported"),
        ("excluded", "included"),
        ("unsafe", "safe"),
        ("crisis", "situation"),
        ("deadly", "alleged"),
        ("unacceptable", "exaggerated")
    ]

    # Ambiguous words to monitor
    ambiguous_words_to_monitor = [
        "systemic", "cultural", "legal", "effective", "necessary",
        "excessive", "problematic", "costly", "specialized", "economic",
        "equitable", "standard", "exceptional", "recognized", "justified",
        "tragic", "constitutional", "complete", "belated", "encouraging",
        "considered", "partial", "disappointing", "bureaucratic", "difficult",
        "commercial", "complex", "modern", "complementary"
    ]

    # Alert thresholds
    bias_threshold_warning = 0.20
    bias_threshold_critical = 0.30

    # CIRBE validation
    cirbe_validation = True
    cirbe_validation_frequency = 1
    cirbe_sample_size = 100

config = ImprovedTrainingConfig()

# Cell 5: Set seeds and device
def set_seed(seed=42):
    np.random.seed(seed)
    torch.manual_seed(seed)
    torch.cuda.manual_seed_all(seed)
    torch.backends.cudnn.deterministic = True
    torch.backends.cudnn.benchmark = False

set_seed(config.random_seed)

# Check GPU
device = torch.device("cuda" if torch.cuda.is_available() else "cpu")
print(f"🖥️ Device: {device}")

if torch.cuda.is_available():
    print(f"🎮 GPU: {torch.cuda.get_device_name(0)}")
    print(f"💾 GPU Memory: {torch.cuda.get_device_properties(0).total_memory / 1024**3:.2f} GB")
    torch.backends.cudnn.benchmark = True
    torch.backends.cuda.matmul.allow_tf32 = True
    os.environ['PYTORCH_CUDA_ALLOC_CONF'] = 'expandable_segments:True'

# Cell 6: Checkpoint functions
def save_checkpoint(model, optimizer, scheduler, scaler, epoch, global_step,
                   training_history, save_path, tokenizer):
    """Save complete checkpoint including all states"""
    os.makedirs(save_path, exist_ok=True)

    # Save model
    model.save_pretrained(save_path)

    # Save tokenizer
    tokenizer.save(os.path.join(save_path, "tokenizer.json"))

    # Save optimizer
    torch.save(optimizer.state_dict(), os.path.join(save_path, "optimizer.pt"))

    # Save scheduler
    torch.save(scheduler.state_dict(), os.path.join(save_path, "scheduler.pt"))

    # Save scaler if used
    if scaler is not None:
        torch.save(scaler.state_dict(), os.path.join(save_path, "scaler.pt"))

    # Save all states
    checkpoint_state = {
        'epoch': epoch,
        'global_step': global_step,
        'random_state': random.getstate(),
        'np_random_state': np.random.get_state(),
        'torch_random_state': torch.get_rng_state(),
        'cuda_random_state': torch.cuda.get_rng_state() if torch.cuda.is_available() else None,
        'training_history': training_history,
        'config': config.__dict__
    }

    torch.save(checkpoint_state, os.path.join(save_path, "training_state.pt"))

    # Also save as JSON for compatibility
    with open(f"{save_path}/training_info.json", 'w') as f:
        json.dump({
            'epoch': epoch,
            'global_step': global_step,
            'config': config.__dict__,
            'history': training_history
        }, f, indent=2)

    print(f"✅ Checkpoint saved: {save_path}")

def load_checkpoint(checkpoint_path, model, optimizer, scheduler, scaler, device):
    """Load complete checkpoint"""
    # Load optimizer
    optimizer_path = os.path.join(checkpoint_path, "optimizer.pt")
    if os.path.exists(optimizer_path):
        optimizer.load_state_dict(torch.load(optimizer_path, map_location=device))
        print("✅ Optimizer state restored")

    # Load scheduler
    scheduler_path = os.path.join(checkpoint_path, "scheduler.pt")
    if os.path.exists(scheduler_path):
        scheduler.load_state_dict(torch.load(scheduler_path, map_location=device))
        print("✅ Learning rate schedule restored")

    # Load scaler
    if scaler is not None:
        scaler_path = os.path.join(checkpoint_path, "scaler.pt")
        if os.path.exists(scaler_path):
            scaler.load_state_dict(torch.load(scaler_path, map_location=device))
            print("✅ FP16 scaler restored")

    # Load states
    state_path = os.path.join(checkpoint_path, "training_state.pt")
    if os.path.exists(state_path):
        checkpoint_state = torch.load(state_path, map_location=device)

        random.setstate(checkpoint_state['random_state'])
        np.random.set_state(checkpoint_state['np_random_state'])
        torch.set_rng_state(checkpoint_state['torch_random_state'])
        if torch.cuda.is_available() and checkpoint_state['cuda_random_state'] is not None:
            torch.cuda.set_rng_state(checkpoint_state['cuda_random_state'])

        print("✅ Random states restored")
        return checkpoint_state
    else:
        # Fallback to JSON
        info_path = os.path.join(checkpoint_path, "training_info.json")
        if os.path.exists(info_path):
            info = safe_load_json(info_path)
            if info is not None:
                return {
                    'epoch': info.get('epoch', 0),
                    'global_step': info.get('global_step', 0),
                    'training_history': info.get('history', {})
                }
    return None

# Cell 7: Mount Drive and paths
from google.colab import drive
drive.mount('/content/drive')

# Define paths
BASE_PATH = "/content/drive/MyDrive/COMP550/final_project"
CANADIAN_HANSARD_PATH = f"{BASE_PATH}/data/canadian_hansard/filtered_canadian_hansard.csv"
NUNAVUT_HANSARD_PATH = f"{BASE_PATH}/data/nunavut_hansard/preprocessed_nunavut_hansard.txt"
CIRBE_PATH = f"{BASE_PATH}/cirbe_bert_ready.csv"
OUTPUT_DIR = f"{BASE_PATH}/models_improved_cirbe"

os.makedirs(OUTPUT_DIR, exist_ok=True)
print(f"✅ Output folder: {OUTPUT_DIR}")

# Cell 8: Load CIRBE
def load_cirbe_benchmark(cirbe_path):
    """Load CIRBE benchmark for validation"""
    print("\n📊 LOADING CIRBE")
    print("=" * 50)

    if os.path.exists(cirbe_path):
        cirbe_df = pd.read_csv(cirbe_path)
        print(f"✅ CIRBE loaded: {len(cirbe_df)} prompts")

        # Analyze categories
        categories = cirbe_df['category'].value_counts()
        print("\n📋 Distribution by category:")
        for cat, count in categories.items():
            print(f"  {cat}: {count} prompts")

        return cirbe_df
    else:
        print(f"❌ CIRBE not found at: {cirbe_path}")
        return None

cirbe_df = load_cirbe_benchmark(CIRBE_PATH)

# Cell 9: Data validation functions
def validate_temporal_alignment(df, expected_start=1999, expected_end=2017):
    """Validate temporal alignment of data"""
    if 'speechdate' not in df.columns:
        print("⚠️ No 'speechdate' column for temporal validation")
        return df

    df['year'] = pd.to_datetime(df['speechdate']).dt.year
    years = df['year'].unique()

    min_year, max_year = years.min(), years.max()
    print(f"📅 Data period: {min_year} - {max_year}")

    if min_year < expected_start or max_year > expected_end:
        df_filtered = df[(df['year'] >= expected_start) & (df['year'] <= expected_end)]
        print(f"✅ Filtering: {len(df)} → {len(df_filtered)} documents")
        return df_filtered
    else:
        print(f"✅ Temporal alignment correct ({expected_start}-{expected_end})")
        return df

def load_and_validate_corpora(canadian_path, nunavut_path):
    """Load and validate corpora"""
    print("\n📚 LOADING AND VALIDATING CORPORA")
    print("=" * 50)

    # Canadian Hansard
    print("\n🇨🇦 Canadian Hansard:")
    canadian_df = pd.read_csv(canadian_path)
    print(f"  Documents loaded: {len(canadian_df)}")

    canadian_df = validate_temporal_alignment(canadian_df)
    canadian_texts = canadian_df['speechtext'].dropna().tolist()

    canadian_lengths = [len(text) for text in canadian_texts]
    print(f"  Valid documents: {len(canadian_texts)}")
    print(f"  Average length: {np.mean(canadian_lengths):.0f} characters")

    # Nunavut Hansard
    print("\n🏔️ Nunavut Hansard:")
    with open(nunavut_path, 'r', encoding='utf-8') as f:
        nunavut_texts = [line.strip() for line in f if line.strip()]

    nunavut_lengths = [len(text) for text in nunavut_texts]
    print(f"  Documents loaded: {len(nunavut_texts)}")
    print(f"  Average length: {np.mean(nunavut_lengths):.0f} characters")

    return canadian_texts, nunavut_texts, canadian_df

# Load data
canadian_texts, nunavut_texts, canadian_df = load_and_validate_corpora(
    CANADIAN_HANSARD_PATH,
    NUNAVUT_HANSARD_PATH
)

# Cell 10: Create datasets
def create_stratified_mixed_datasets(canadian_texts, nunavut_texts, canadian_df=None):
    """Create datasets with stratified random sampling"""
    print("\n🎲 CREATING STRATIFIED DATASETS")
    print("=" * 50)

    np.random.seed(config.random_seed)
    datasets = {}

    min_size = min(len(canadian_texts), len(nunavut_texts))
    print(f"Target size per dataset: {min_size} documents")

    # Random indices
    canadian_indices = np.random.permutation(len(canadian_texts))[:min_size]
    nunavut_indices = np.random.permutation(len(nunavut_texts))[:min_size]

    # Pure datasets
    datasets['canadian_only'] = {
        'texts': [canadian_texts[i] for i in canadian_indices],
        'indices': canadian_indices.tolist(),
        'source': 'canadian'
    }

    datasets['nunavut_only'] = {
        'texts': [nunavut_texts[i] for i in nunavut_indices],
        'indices': nunavut_indices.tolist(),
        'source': 'nunavut'
    }

    # Mixed datasets
    mixing_strategies = [
        ('imbalanced_80_20', 0.8),
        ('balanced_50_50', 0.5),
        ('imbalanced_20_80', 0.2)
    ]

    for name, canadian_ratio in mixing_strategies:
        n_canadian = int(min_size * canadian_ratio)
        n_nunavut = min_size - n_canadian

        selected_canadian_idx = np.random.choice(canadian_indices, n_canadian, replace=False)
        selected_nunavut_idx = np.random.choice(nunavut_indices, n_nunavut, replace=False)

        mixed_texts = []
        mixed_sources = []

        for idx in selected_canadian_idx:
            mixed_texts.append(canadian_texts[idx])
            mixed_sources.append('canadian')

        for idx in selected_nunavut_idx:
            mixed_texts.append(nunavut_texts[idx])
            mixed_sources.append('nunavut')

        shuffle_idx = np.random.permutation(len(mixed_texts))
        mixed_texts = [mixed_texts[i] for i in shuffle_idx]
        mixed_sources = [mixed_sources[i] for i in shuffle_idx]

        datasets[name] = {
            'texts': mixed_texts,
            'sources': mixed_sources,
            'canadian_ratio': canadian_ratio
        }

        print(f"\n✅ {name}:")
        print(f"   - Canadian: {n_canadian} ({canadian_ratio*100:.0f}%)")
        print(f"   - Nunavut: {n_nunavut} ({(1-canadian_ratio)*100:.0f}%)")

    return datasets

datasets = create_stratified_mixed_datasets(canadian_texts, nunavut_texts, canadian_df)

# Cell 11: Tokenizer
tokenizer_path = f"{OUTPUT_DIR}/tokenizer/tokenizer.json"

if os.path.exists(tokenizer_path):
    print("✅ Loading existing tokenizer")
    tokenizer = Tokenizer.from_file(tokenizer_path)
    print(f"  Vocabulary size: {tokenizer.get_vocab_size()}")
else:
    print("🔨 Training new tokenizer")
    all_texts = canadian_texts + nunavut_texts
    np.random.shuffle(all_texts)

    tokenizer = BertWordPieceTokenizer(
        clean_text=True,
        handle_chinese_chars=False,
        strip_accents=False,
        lowercase=False
    )

    # Training corpus
    temp_file = "temp_corpus_tokenizer.txt"
    sample_size = min(100000, len(all_texts))
    sample_indices = np.random.choice(len(all_texts), sample_size, replace=False)

    with open(temp_file, 'w', encoding='utf-8') as f:
        for idx in sample_indices:
            f.write(all_texts[idx] + '\n')

    # Special tokens
    special_tokens = ["[PAD]", "[UNK]", "[CLS]", "[SEP]", "[MASK]"]

    print(f"Training on {sample_size} documents...")
    tokenizer.train(
        files=[temp_file],
        vocab_size=config.vocab_size,
        min_frequency=2,
        special_tokens=special_tokens,
        limit_alphabet=1000,
        wordpieces_prefix="##"
    )

    os.remove(temp_file)

    tokenizer.post_processor = BertProcessing(
        ("[SEP]", tokenizer.token_to_id("[SEP]")),
        ("[CLS]", tokenizer.token_to_id("[CLS]"))
    )

    os.makedirs(os.path.dirname(tokenizer_path), exist_ok=True)
    tokenizer.save(tokenizer_path)
    print(f"✅ Tokenizer saved: {tokenizer_path}")

print(f"📊 Vocabulary size: {tokenizer.get_vocab_size()}")

# Cell 12: Synonym Handler
class ReproducibleSynonymHandler:
    """Synonym handler using external sources for reproducibility"""

    def __init__(self, tokenizer, cache_dir="./synonym_cache"):
        self.tokenizer = tokenizer
        self.cache_dir = cache_dir
        os.makedirs(cache_dir, exist_ok=True)

        self._initialize_wordnet()
        self._load_conceptnet()
        self.domain_specific_synonyms = self._load_domain_specific_synonyms()

    def _initialize_wordnet(self):
        """Download and initialize WordNet if necessary"""
        try:
            nltk.data.find('corpora/wordnet')
            print("✅ WordNet already downloaded")
        except LookupError:
            print("📥 Downloading WordNet...")
            nltk.download('wordnet')
            nltk.download('omw-1.4')

    def _load_conceptnet(self):
        """Load ConceptNet data"""
        self.conceptnet_cache = os.path.join(self.cache_dir, "conceptnet_synonyms.json")

        if os.path.exists(self.conceptnet_cache):
            with open(self.conceptnet_cache, 'r') as f:
                self.conceptnet_data = json.load(f)
            print("✅ ConceptNet cache loaded")
        else:
            print("📥 ConceptNet cache not found")
            self.conceptnet_data = {}

    def _load_domain_specific_synonyms(self) -> Dict[str, List[str]]:
        """Load domain-specific synonyms"""
        domain_file = os.path.join(self.cache_dir, "indigenous_rights_synonyms.json")

        if os.path.exists(domain_file):
            with open(domain_file, 'r') as f:
                return json.load(f)
        else:
            base_mappings = {
                "systemic": ["systematic", "structural", "institutional", "pervasive"],
                "discriminatory": ["biased", "prejudicial", "unfair", "inequitable"],
                "insufficient": ["inadequate", "deficient", "lacking", "scarce"],
                "underfunded": ["underfinanced", "underresourced", "deprived"],
                "violated": ["breached", "infringed", "broken", "transgressed"],
                "temporary": ["transient", "brief", "short-term", "momentary"],
                "individual": ["isolated", "personal", "singular", "specific"],
                "adequate": ["sufficient", "satisfactory", "acceptable", "suitable"],
                "generous": ["ample", "abundant", "substantial", "plentiful"],
                "Indigenous": ["Aboriginal", "Native", "First Peoples"],
                "First Nations": ["Native", "Aboriginal"],
                "Métis": [],
                "Inuit": []
            }

            with open(domain_file, 'w') as f:
                json.dump(base_mappings, f, indent=2)

            print(f"✅ Domain synonyms file created: {domain_file}")
            return base_mappings

    def get_wordnet_synonyms(self, word: str) -> Set[str]:
        """Get synonyms from WordNet"""
        synonyms = set()

        for syn in wordnet.synsets(word):
            for lemma in syn.lemmas():
                synonym = lemma.name().replace('_', ' ')
                if synonym.lower() != word.lower():
                    synonyms.add(synonym)

        return synonyms

    def find_validated_synonym(self, word: str, context: str = "general") -> Optional[str]:
        """Find a validated synonym"""
        if self.tokenizer.token_to_id(word) is not None:
            return word

        if self.tokenizer.token_to_id(word.lower()) is not None:
            return word.lower()

        # Check domain-specific synonyms
        if word in self.domain_specific_synonyms:
            for synonym in self.domain_specific_synonyms[word]:
                if self.tokenizer.token_to_id(synonym) is not None:
                    return synonym
                if self.tokenizer.token_to_id(synonym.lower()) is not None:
                    return synonym.lower()

        # Check WordNet
        wordnet_synonyms = self.get_wordnet_synonyms(word)
        for synonym in wordnet_synonyms:
            if self.tokenizer.token_to_id(synonym) is not None:
                return synonym
            if self.tokenizer.token_to_id(synonym.lower()) is not None:
                return synonym.lower()

        return None

# Cell 13: CIRBE validation with synonyms
def validate_cirbe_with_reproducible_synonyms(cirbe_df, tokenizer, output_dir):
    """Validate CIRBE with synonyms"""
    os.makedirs(output_dir, exist_ok=True)

    synonym_handler = ReproducibleSynonymHandler(
        tokenizer,
        cache_dir=os.path.join(output_dir, "synonym_cache")
    )

    all_words = set()
    all_words.update(cirbe_df['favorable_answer'].unique())
    all_words.update(cirbe_df['unfavorable_answer'].unique())

    print(f"\n📊 Analyzing {len(all_words)} unique words in CIRBE...")

    fixed_cirbe = cirbe_df.copy()
    modifications = []

    for idx, row in fixed_cirbe.iterrows():
        # Check favorable_answer
        if tokenizer.token_to_id(row['favorable_answer']) is None:
            synonym = synonym_handler.find_validated_synonym(row['favorable_answer'])
            if synonym and synonym != row['favorable_answer']:
                modifications.append({
                    'prompt_idx': idx,
                    'field': 'favorable_answer',
                    'original': row['favorable_answer'],
                    'replacement': synonym
                })
                fixed_cirbe.at[idx, 'favorable_answer'] = synonym

        # Check unfavorable_answer
        if tokenizer.token_to_id(row['unfavorable_answer']) is None:
            synonym = synonym_handler.find_validated_synonym(row['unfavorable_answer'])
            if synonym and synonym != row['unfavorable_answer']:
                modifications.append({
                    'prompt_idx': idx,
                    'field': 'unfavorable_answer',
                    'original': row['unfavorable_answer'],
                    'replacement': synonym
                })
                fixed_cirbe.at[idx, 'unfavorable_answer'] = synonym

    # Save modifications
    if modifications:
        modifications_df = pd.DataFrame(modifications)
        modifications_path = os.path.join(output_dir, "cirbe_modifications.csv")
        modifications_df.to_csv(modifications_path, index=False)
        print(f"\n📝 {len(modifications)} modifications applied")

    return fixed_cirbe, modifications

# Validate CIRBE
if cirbe_df is not None:
    print("\n🔧 VALIDATING CIRBE WITH REPRODUCIBLE SYNONYMS")
    print("=" * 70)

    cirbe_validation_dir = os.path.join(OUTPUT_DIR, "cirbe_validation")
    cirbe_df_fixed, modifications_df = validate_cirbe_with_reproducible_synonyms(
        cirbe_df,
        tokenizer,
        output_dir=cirbe_validation_dir
    )

    cirbe_df = cirbe_df_fixed
    print(f"✅ CIRBE ready with {len(cirbe_df)} valid prompts")

# Cell 14: Bias Monitor
class ImprovedBiasMonitor:
    """Monitor biases during training"""

    def __init__(self, word_pairs, ambiguous_words, tokenizer,
                 threshold_warning=0.20, threshold_critical=0.30):
        self.word_pairs = word_pairs
        self.ambiguous_words = ambiguous_words
        self.tokenizer = tokenizer
        self.threshold_warning = threshold_warning
        self.threshold_critical = threshold_critical
        self.history = []

    def compute_bias_scores(self, model):
        """Calculate bias scores"""
        model.eval()
        embeddings = model.bert.embeddings.word_embeddings.weight.data

        scores = {}
        alerts = []
        ambiguous_analysis = {}

        with torch.no_grad():
            for fav_word, unfav_word in self.word_pairs:
                fav_id = self.tokenizer.token_to_id(fav_word)
                unfav_id = self.tokenizer.token_to_id(unfav_word)

                if fav_id is not None and unfav_id is not None:
                    emb_fav = embeddings[fav_id].cpu().numpy()
                    emb_unfav = embeddings[unfav_id].cpu().numpy()

                    similarity = cosine_similarity([emb_fav], [emb_unfav])[0][0]
                    pair_name = f"{fav_word}-{unfav_word}"
                    scores[pair_name] = similarity

                    if similarity > self.threshold_critical:
                        alerts.append(
                            f"🚨 CRITICAL: '{fav_word}' and '{unfav_word}' "
                            f"too similar ({similarity:.3f})"
                        )
                    elif similarity > self.threshold_warning:
                        alerts.append(
                            f"⚠️ Warning: '{fav_word}' and '{unfav_word}' "
                            f"high similarity ({similarity:.3f})"
                        )

        return scores, alerts, ambiguous_analysis

    def check_cirbe_alignment(self, model, cirbe_prompts_sample):
        """Check alignment with CIRBE expectations"""
        model.eval()
        correct = 0
        total = 0

        with torch.no_grad():
            for _, row in cirbe_prompts_sample.iterrows():
                prompt = row['prompt']
                favorable = row['favorable_answer']
                unfavorable = row['unfavorable_answer']

                encoding = self.tokenizer.encode(prompt)
                input_ids = torch.tensor([encoding.ids]).to(device)

                mask_token_id = self.tokenizer.token_to_id('[MASK]')
                mask_positions = (input_ids == mask_token_id).nonzero(as_tuple=True)

                if len(mask_positions[1]) == 0:
                    continue

                mask_idx = mask_positions[1][0]

                outputs = model(input_ids)
                predictions = outputs.logits[0, mask_idx]

                fav_id = self.tokenizer.token_to_id(favorable)
                unfav_id = self.tokenizer.token_to_id(unfavorable)

                if fav_id is not None and unfav_id is not None:
                    if predictions[fav_id] > predictions[unfav_id]:
                        correct += 1
                    total += 1

        accuracy = correct / total if total > 0 else 0
        return accuracy, correct, total

    def log_bias_check(self, epoch, step, scores, alerts, cirbe_accuracy=None):
        """Log results with CIRBE score"""
        self.history.append({
            'epoch': epoch,
            'step': step,
            'scores': scores,
            'alerts': alerts,
            'cirbe_accuracy': cirbe_accuracy,
            'timestamp': datetime.now()
        })

        if alerts:
            print(f"\n{'='*50}")
            print(f"BIAS ALERT - Epoch {epoch}, Step {step}")
            print(f"{'='*50}")
            for alert in alerts:
                print(alert)
            if cirbe_accuracy is not None:
                print(f"\n📊 CIRBE Accuracy: {cirbe_accuracy:.2%}")
            print(f"{'='*50}\n")

    def plot_bias_evolution(self, save_path=None):
        """Visualize bias evolution"""
        if not self.history:
            return

        fig, (ax1, ax2) = plt.subplots(2, 1, figsize=(12, 10))

        # Plot bias evolution
        steps = [h['step'] for h in self.history]
        pair_scores = {}

        for pair in self.word_pairs:
            pair_name = f"{pair[0]}-{pair[1]}"
            scores = [h['scores'].get(pair_name, 0) for h in self.history]
            if any(s != 0 for s in scores):
                pair_scores[pair_name] = scores

        for pair_name, scores in pair_scores.items():
            ax1.plot(steps, scores, label=pair_name, marker='o', markersize=4)

        ax1.axhline(y=self.threshold_warning, color='orange', linestyle='--', alpha=0.5)
        ax1.axhline(y=self.threshold_critical, color='red', linestyle='--', alpha=0.5)
        ax1.set_xlabel('Steps')
        ax1.set_ylabel('Cosine Similarity')
        ax1.set_title('Bias Evolution During Training')
        ax1.legend(bbox_to_anchor=(1.05, 1), loc='upper left')
        ax1.grid(True, alpha=0.3)

        # Plot CIRBE accuracy
        cirbe_steps = [h['step'] for h in self.history if h.get('cirbe_accuracy') is not None]
        cirbe_accs = [h['cirbe_accuracy'] for h in self.history if h.get('cirbe_accuracy') is not None]

        if cirbe_accs:
            ax2.plot(cirbe_steps, cirbe_accs, 'g-', marker='s', markersize=6, linewidth=2)
            ax2.axhline(y=0.5, color='gray', linestyle='--', alpha=0.5, label='Random (50%)')
            ax2.set_xlabel('Steps')
            ax2.set_ylabel('CIRBE Accuracy')
            ax2.set_title('CIRBE Alignment During Training')
            ax2.set_ylim(0, 1)
            ax2.grid(True, alpha=0.3)
            ax2.legend()

        plt.tight_layout()

        if save_path:
            plt.savefig(save_path, dpi=300, bbox_inches='tight')
        plt.show()

# Create bias monitor
bias_monitor = ImprovedBiasMonitor(
    config.bias_word_pairs,
    config.ambiguous_words_to_monitor,
    tokenizer,
    config.bias_threshold_warning,
    config.bias_threshold_critical
)

# Cell 15: Dataset class
class ImprovedBERTDataset(Dataset):
    """Dataset with source tracking"""

    def __init__(self, texts, tokenizer, max_length=512, sources=None):
        self.texts = texts
        self.tokenizer = tokenizer
        self.max_length = max_length
        self.sources = sources if sources else ['unknown'] * len(texts)

    def __len__(self):
        return len(self.texts)

    def __getitem__(self, idx):
        text = self.texts[idx]
        source = self.sources[idx] if idx < len(self.sources) else 'unknown'

        encoding = self.tokenizer.encode(text)

        if len(encoding.ids) > self.max_length - 2:
            ids = encoding.ids[:self.max_length - 2]
            ids = [self.tokenizer.token_to_id("[CLS]")] + ids + [self.tokenizer.token_to_id("[SEP]")]
        else:
            ids = encoding.ids

        padding_length = self.max_length - len(ids)
        attention_mask = [1] * len(ids) + [0] * padding_length
        ids = ids + [self.tokenizer.token_to_id("[PAD]")] * padding_length

        return {
            'input_ids': torch.tensor(ids, dtype=torch.long),
            'attention_mask': torch.tensor(attention_mask, dtype=torch.long),
            'source': source
        }

# Cell 16: Training function
def train_model_with_bias_monitoring(
    model_name,
    dataset_info,
    tokenizer,
    num_epochs=15,
    output_dir="./models_improved",
    bias_monitor=None,
    cirbe_df=None,
    resume_from_checkpoint=None
):
    """Train model with bias monitoring and CIRBE validation"""

    print(f"\n{'='*70}")
    print(f"TRAINING: {model_name}")
    print(f"Dataset: {len(dataset_info['texts'])} documents")
    if 'canadian_ratio' in dataset_info:
        print(f"Canadian ratio: {dataset_info['canadian_ratio']*100:.0f}%")
    print(f"{'='*70}")

    # Check existing checkpoints
    checkpoint_dir = f"{output_dir}/{model_name}"
    checkpoints = []
    if os.path.exists(checkpoint_dir):
        checkpoints = [d for d in os.listdir(checkpoint_dir) if d.startswith('checkpoint-')]
        checkpoints.sort(key=lambda x: int(x.split('-')[1]))

    # Initial state
    start_epoch = 0
    global_step = 0
    best_val_loss = float('inf')
    best_cirbe_accuracy = 0.0
    training_history = {
        'train_losses': [],
        'val_losses': [],
        'bias_checks': [],
        'learning_rates': [],
        'cirbe_accuracies': []
    }

    # Resume from checkpoint
    if resume_from_checkpoint == "latest" and checkpoints:
        resume_from_checkpoint = checkpoints[-1]

    # Create or load model
    vocab_size = tokenizer.get_vocab_size()
    model = None

    if resume_from_checkpoint and checkpoints:
        checkpoint_path = f"{checkpoint_dir}/{resume_from_checkpoint}"
        if os.path.exists(checkpoint_path):
            print(f"\n📂 RESUMING FROM: {resume_from_checkpoint}")
            model = BertForMaskedLM.from_pretrained(checkpoint_path)
            model.to(device)
            print("✅ Model loaded from checkpoint")

    # Create new model if necessary
    if model is None:
        print("\n🆕 Starting from scratch")
        bert_config = BertConfig(
            vocab_size=vocab_size,
            hidden_size=768,
            num_hidden_layers=12,
            num_attention_heads=12,
            intermediate_size=3072,
            hidden_act="gelu",
            hidden_dropout_prob=0.1,
            attention_probs_dropout_prob=0.1,
            max_position_embeddings=512,
            type_vocab_size=2,
            initializer_range=0.02,
            layer_norm_eps=1e-12,
        )

        model = BertForMaskedLM(bert_config)
        model.to(device)

    # Mixed precision
    scaler = GradScaler() if config.fp16 else None
    if config.fp16:
        print("✅ FP16 enabled")

    # Dataset
    sources = dataset_info.get('sources', ['unknown'] * len(dataset_info['texts']))
    dataset = ImprovedBERTDataset(
        dataset_info['texts'],
        tokenizer,
        config.max_length,
        sources
    )

    train_size = int((1 - config.validation_split) * len(dataset))
    val_size = len(dataset) - train_size
    train_dataset, val_dataset = random_split(dataset, [train_size, val_size])

    print(f"\n📊 Split:")
    print(f"  Train: {train_size}")
    print(f"  Val: {val_size}")

    # Prepare CIRBE sample
    cirbe_sample = None
    if cirbe_df is not None and config.cirbe_validation:
        cirbe_sample = cirbe_df.sample(
            n=min(config.cirbe_sample_size, len(cirbe_df)),
            random_state=config.random_seed
        )
        print(f"  CIRBE sample: {len(cirbe_sample)} prompts")

    # DataLoader collate function
    def collate_fn(batch):
        input_ids = torch.stack([item['input_ids'] for item in batch])
        attention_mask = torch.stack([item['attention_mask'] for item in batch])
        sources = [item['source'] for item in batch]

        # MLM
        labels = input_ids.clone()
        probability_matrix = torch.full(labels.shape, config.mlm_probability)

        special_tokens_mask = (
            (input_ids == tokenizer.token_to_id("[PAD]")) |
            (input_ids == tokenizer.token_to_id("[CLS]")) |
            (input_ids == tokenizer.token_to_id("[SEP]"))
        )
        probability_matrix.masked_fill_(special_tokens_mask, value=0.0)

        masked_indices = torch.bernoulli(probability_matrix).bool()
        labels[~masked_indices] = -100

        indices_replaced = torch.bernoulli(torch.full(labels.shape, 0.8)).bool() & masked_indices
        input_ids[indices_replaced] = tokenizer.token_to_id("[MASK]")

        indices_random = torch.bernoulli(torch.full(labels.shape, 0.5)).bool() & masked_indices & ~indices_replaced
        random_words = torch.randint(tokenizer.get_vocab_size(), labels.shape, dtype=torch.long)
        input_ids[indices_random] = random_words[indices_random]

        return {
            'input_ids': input_ids,
            'attention_mask': attention_mask,
            'labels': labels,
            'sources': sources
        }

    # DataLoaders
    train_loader = DataLoader(
        train_dataset,
        batch_size=config.batch_size,
        shuffle=True,
        collate_fn=collate_fn,
        num_workers=2,
        pin_memory=True
    )

    val_loader = DataLoader(
        val_dataset,
        batch_size=config.batch_size * 2,
        collate_fn=collate_fn,
        num_workers=2,
        pin_memory=True
    )

    # Optimizer and Scheduler
    optimizer = torch.optim.AdamW(
        model.parameters(),
        lr=config.learning_rate,
        weight_decay=config.weight_decay
    )

    total_steps = len(train_loader) * num_epochs
    scheduler = get_linear_schedule_with_warmup(
        optimizer,
        num_warmup_steps=config.warmup_steps,
        num_training_steps=total_steps
    )

    # Load checkpoint states
    if resume_from_checkpoint and checkpoints:
        checkpoint_path = f"{checkpoint_dir}/{resume_from_checkpoint}"
        if os.path.exists(checkpoint_path):
            checkpoint_state = load_checkpoint(checkpoint_path, model, optimizer, scheduler, scaler, device)
            if checkpoint_state:
                start_epoch = checkpoint_state.get('epoch', 0) + 1
                global_step = checkpoint_state.get('global_step', 0)
                training_history = checkpoint_state.get('training_history', training_history)
                print(f"✅ Resuming from epoch {start_epoch}, step {global_step}")

    # Variables for early stopping
    patience = 5
    patience_counter = 0

    # Training loop
    for epoch in range(start_epoch, num_epochs):
        # Training
        model.train()
        epoch_loss = 0
        progress_bar = tqdm(train_loader, desc=f"Epoch {epoch+1}/{num_epochs}")

        for step, batch in enumerate(progress_bar):
            input_ids = batch['input_ids'].to(device)
            attention_mask = batch['attention_mask'].to(device)
            labels = batch['labels'].to(device)

            # Forward pass
            if config.fp16 and scaler is not None:
                with autocast():
                    outputs = model(
                        input_ids=input_ids,
                        attention_mask=attention_mask,
                        labels=labels
                    )
                    loss = outputs.loss / config.gradient_accumulation_steps

                scaler.scale(loss).backward()
            else:
                outputs = model(
                    input_ids=input_ids,
                    attention_mask=attention_mask,
                    labels=labels
                )
                loss = outputs.loss / config.gradient_accumulation_steps
                loss.backward()

            epoch_loss += loss.item()

            # Gradient accumulation
            if (step + 1) % config.gradient_accumulation_steps == 0:
                if config.fp16 and scaler is not None:
                    scaler.unscale_(optimizer)
                    torch.nn.utils.clip_grad_norm_(model.parameters(), config.max_grad_norm)
                    scaler.step(optimizer)
                    scaler.update()
                else:
                    torch.nn.utils.clip_grad_norm_(model.parameters(), config.max_grad_norm)
                    optimizer.step()

                scheduler.step()
                optimizer.zero_grad()
                global_step += 1

                # Logging
                if global_step % config.logging_steps == 0:
                    avg_loss = epoch_loss / (step + 1) * config.gradient_accumulation_steps
                    current_lr = scheduler.get_last_lr()[0]

                    progress_bar.set_postfix({
                        'loss': f'{avg_loss:.4f}',
                        'lr': f'{current_lr:.2e}'
                    })

                    training_history['train_losses'].append(avg_loss)
                    training_history['learning_rates'].append(current_lr)

                # Bias monitoring
                if bias_monitor and global_step % config.bias_check_frequency == 0:
                    scores, alerts, ambiguous_analysis = bias_monitor.compute_bias_scores(model)

                    cirbe_acc = None
                    if cirbe_sample is not None:
                        cirbe_acc, _, _ = bias_monitor.check_cirbe_alignment(model, cirbe_sample)
                        training_history['cirbe_accuracies'].append({
                            'step': global_step,
                            'accuracy': cirbe_acc
                        })

                    bias_monitor.log_bias_check(epoch, global_step, scores, alerts, cirbe_acc)
                    training_history['bias_checks'].append({
                        'step': global_step,
                        'scores': scores,
                        'alerts': alerts,
                        'ambiguous_analysis': ambiguous_analysis
                    })

                    # Early stopping if biases too strong
                    if config.bias_early_stopping:
                        critical_count = sum(1 for a in alerts if "CRITICAL" in a)
                        if critical_count > 3:
                            print(f"\n🛑 Stopping: too many critical biases ({critical_count})")
                            return model, training_history

                # Save checkpoint
                if global_step % config.save_steps == 0:
                    save_path = f"{output_dir}/{model_name}/checkpoint-{global_step}"
                    save_checkpoint(model, optimizer, scheduler, scaler, epoch, global_step,
                                  training_history, save_path, tokenizer)

        # Validation
        model.eval()
        val_loss = 0

        with torch.no_grad():
            for batch in tqdm(val_loader, desc="Validation"):
                input_ids = batch['input_ids'].to(device)
                attention_mask = batch['attention_mask'].to(device)
                labels = batch['labels'].to(device)

                outputs = model(
                    input_ids=input_ids,
                    attention_mask=attention_mask,
                    labels=labels
                )

                val_loss += outputs.loss.item()

        avg_val_loss = val_loss / len(val_loader)
        training_history['val_losses'].append(avg_val_loss)

        # CIRBE validation
        cirbe_epoch_accuracy = 0.0
        if cirbe_sample is not None and config.cirbe_validation:
            print("\n🎯 CIRBE Validation...")
            cirbe_epoch_accuracy, correct, total = bias_monitor.check_cirbe_alignment(model, cirbe_sample)
            print(f"  Accuracy: {cirbe_epoch_accuracy:.2%} ({correct}/{total})")

            if cirbe_epoch_accuracy > best_cirbe_accuracy:
                best_cirbe_accuracy = cirbe_epoch_accuracy
                print(f"  ✨ New best CIRBE score!")

        print(f"\n📊 Epoch {epoch+1}:")
        print(f"  Train Loss: {epoch_loss/len(train_loader):.4f}")
        print(f"  Val Loss: {avg_val_loss:.4f}")
        if cirbe_epoch_accuracy > 0:
            print(f"  CIRBE Accuracy: {cirbe_epoch_accuracy:.2%}")

        # Early stopping
        if avg_val_loss < best_val_loss:
            best_val_loss = avg_val_loss
            patience_counter = 0

            best_path = f"{output_dir}/{model_name}/best_model"
            os.makedirs(best_path, exist_ok=True)
            model.save_pretrained(best_path)
            tokenizer.save(os.path.join(best_path, "tokenizer.json"))

            with open(f"{best_path}/best_metrics.json", 'w') as f:
                json.dump({
                    'val_loss': best_val_loss,
                    'cirbe_accuracy': cirbe_epoch_accuracy,
                    'epoch': epoch
                }, f, indent=2)

            print(f"✨ Best model saved (loss: {best_val_loss:.4f})")
        else:
            patience_counter += 1
            if patience_counter >= patience:
                print(f"\n⚠️ Early stopping (patience: {patience})")
                break

    # Save final model
    final_path = f"{output_dir}/{model_name}/final_model"
    os.makedirs(final_path, exist_ok=True)
    model.save_pretrained(final_path)
    tokenizer.save(os.path.join(final_path, "tokenizer.json"))

    with open(f"{final_path}/final_metrics.json", 'w') as f:
        json.dump({
            'best_val_loss': best_val_loss,
            'best_cirbe_accuracy': best_cirbe_accuracy,
            'final_epoch': epoch,
            'total_steps': global_step
        }, f, indent=2)

    # Bias report
    if bias_monitor:
        bias_monitor.plot_bias_evolution(
            save_path=f"{output_dir}/{model_name}/bias_evolution.png"
        )

    return model, training_history

# Cell 17: Main training loop
models_to_train = {
    'canadian_only': datasets['canadian_only'],
    'imbalanced_80_20': datasets['imbalanced_80_20'],
    'balanced_50_50': datasets['balanced_50_50'],
    'imbalanced_20_80': datasets['imbalanced_20_80'],
    'nunavut_only': datasets['nunavut_only']
}

print("🎯 Models to train:")
for name in models_to_train:
    print(f"  - {name}")

# Train models
results = {}

for model_name, dataset_info in models_to_train.items():
    print(f"\n{'='*70}")
    print(f"🚀 STARTING: {model_name}")
    print(f"{'='*70}")

    # Check if we should resume
    should_resume = False
    if os.path.exists(OUTPUT_DIR):
        model_dir = os.path.join(OUTPUT_DIR, model_name)
        if os.path.exists(model_dir):
            checkpoints = [d for d in os.listdir(model_dir) if d.startswith('checkpoint-')]
            if checkpoints:
                should_resume = True
                print(f"🔄 Checkpoints found, resume possible")

    try:
        model, history = train_model_with_bias_monitoring(
            model_name,
            dataset_info,
            tokenizer,
            num_epochs=config.num_epochs,
            output_dir=OUTPUT_DIR,
            bias_monitor=bias_monitor,
            cirbe_df=cirbe_df,
            resume_from_checkpoint="latest" if should_resume else None
        )

        results[model_name] = {
            'status': 'success',
            'history': history,
            'final_val_loss': history['val_losses'][-1] if history['val_losses'] else None,
            'best_cirbe_accuracy': max([ca['accuracy'] for ca in history.get('cirbe_accuracies', [{'accuracy': 0}])], default=0)
        }

        # Clean memory
        del model
        torch.cuda.empty_cache()

    except Exception as e:
        print(f"\n❌ Error: {e}")
        import traceback
        traceback.print_exc()

        results[model_name] = {
            'status': 'failed',
            'error': str(e)
        }

    # Pause between models
    time.sleep(10)

# Cell 18: Final report
print("\n" + "="*70)
print("📊 FINAL REPORT WITH CIRBE ANALYSIS")
print("="*70)

# Summary table
summary_data = []

for model_name, result in results.items():
    if result['status'] == 'success':
        summary_data.append({
            'Model': model_name,
            'Status': '✅',
            'Final Val Loss': f"{result['final_val_loss']:.4f}" if result['final_val_loss'] else 'N/A',
            'Best CIRBE Acc': f"{result.get('best_cirbe_accuracy', 0):.2%}"
        })
    else:
        summary_data.append({
            'Model': model_name,
            'Status': '❌',
            'Final Val Loss': 'Failed',
            'Best CIRBE Acc': 'Failed'
        })

# Display table
summary_df = pd.DataFrame(summary_data)
print("\n", summary_df.to_string(index=False))

# Save complete report
report_path = f"{OUTPUT_DIR}/training_report_complete.json"
with open(report_path, 'w') as f:
    json.dump({
        'results': results,
        'config': config.__dict__,
        'summary': summary_data,
        'timestamp': datetime.now().isoformat()
    }, f, indent=2)

print(f"\n✅ Training completed!")
print(f"📁 Models saved in: {OUTPUT_DIR}")
print(f"📊 Complete report: {report_path}")

# Visualize bias evolution
if bias_monitor and bias_monitor.history:
    print("\n📈 Bias evolution and CIRBE alignment:")
    bias_monitor.plot_bias_evolution(
        save_path=f"{OUTPUT_DIR}/bias_cirbe_evolution_complete.png"
    )

print("\n🎉 TRAINING PIPELINE COMPLETED!")
print("Your work contributes to more ethical AI development. 🌟")
```

```python
#!/usr/bin/env python3
"""
Create alternate CIRBE benchmark where ALL favorable answers
are replaced with Nunavut model's top prediction.

This creates a benchmark that fully represents what the Nunavut
digital expert considers the most likely/appropriate completions.
"""

import os
import torch
import pandas as pd
import numpy as np
from tqdm import tqdm
from transformers import BertForMaskedLM
from tokenizers import Tokenizer
from datetime import datetime
import json
import matplotlib.pyplot as plt
import seaborn as sns

# Configuration
BASE_PATH = "/content/drive/MyDrive/COMP550/final_project"
MODELS_DIR = os.path.join(BASE_PATH, "models_improved_cirbe")
NUNAVUT_MODEL_PATH = os.path.join(MODELS_DIR, "nunavut_only/best_model")
CIRBE_PATH = os.path.join(BASE_PATH, "cirbe_bert_ready.csv")
OUTPUT_DIR = os.path.join(BASE_PATH, "cirbe_nunavut_full_validation")

# Create output directory
os.makedirs(OUTPUT_DIR, exist_ok=True)

# Device setup
device = torch.device("cuda" if torch.cuda.is_available() else "cpu")
print(f"🖥️ Using device: {device}")

def load_model_and_tokenizer():
    """Load the Nunavut-only model and tokenizer"""
    print("\n📚 Loading Nunavut Digital Expert Model...")

    # Load tokenizer
    tokenizer_path = os.path.join(NUNAVUT_MODEL_PATH, "tokenizer.json")
    tokenizer = Tokenizer.from_file(tokenizer_path)
    print(f"✅ Tokenizer loaded (vocab size: {tokenizer.get_vocab_size()})")

    # Load model
    model = BertForMaskedLM.from_pretrained(NUNAVUT_MODEL_PATH)
    model.to(device)
    model.eval()
    print(f"✅ Model loaded and ready on {device}")

    return tokenizer, model

def get_top_prediction(prompt, model, tokenizer):
    """Get the #1 prediction for a masked prompt"""

    # Ensure consistent mask token format
    prompt_fixed = prompt.replace('[MASK]', '[MASK]')

    # Tokenize
    encoding = tokenizer.encode(prompt_fixed)
    input_ids = torch.tensor([encoding.ids]).to(device)

    # Find mask position
    mask_token_id = tokenizer.token_to_id("[MASK]")
    mask_positions = (input_ids == mask_token_id).nonzero(as_tuple=True)

    if len(mask_positions[1]) == 0:
        return None, None, "No mask token found"

    mask_idx = mask_positions[1][0].item()

    # Get predictions
    with torch.no_grad():
        outputs = model(input_ids=input_ids)
        predictions = outputs.logits[0, mask_idx]
        probabilities = torch.softmax(predictions, dim=-1)

    # Get top prediction
    top_prob, top_idx = torch.max(probabilities, dim=0)
    top_token = tokenizer.decode([top_idx.item()])

    # Clean token
    top_token_clean = top_token.replace('##', '').strip()

    # Also get top 5 for analysis
    top5_probs, top5_indices = torch.topk(probabilities, 5)
    top5_tokens = []
    for prob, idx in zip(top5_probs, top5_indices):
        token = tokenizer.decode([idx.item()]).replace('##', '').strip()
        if token:
            top5_tokens.append((token, prob.item()))

    return top_token_clean, top_prob.item(), top5_tokens

def create_nunavut_validated_benchmark(cirbe_df, model, tokenizer):
    """Replace ALL favorable answers with Nunavut's top predictions"""

    print(f"\n🔄 Replacing ALL favorable answers with Nunavut predictions...")
    print("This creates a benchmark fully grounded in Inuit discourse\n")

    results = []

    for idx, row in tqdm(cirbe_df.iterrows(), total=len(cirbe_df), desc="Processing"):
        prompt = row['prompt']
        original_favorable = row['favorable_answer']
        original_unfavorable = row['unfavorable_answer']
        category = row.get('category', 'Unknown')

        # Get Nunavut's top prediction
        nunavut_answer, confidence, top5 = get_top_prediction(prompt, model, tokenizer)

        if nunavut_answer is None:
            # Failed to get prediction
            results.append({
                'prompt_id': idx,
                'prompt': prompt,
                'category': category,
                'original_favorable': original_favorable,
                'original_unfavorable': original_unfavorable,
                'nunavut_favorable': original_favorable,  # Keep original if failed
                'confidence': 0.0,
                'status': 'Failed',
                'top_5_predictions': ''
            })
        else:
            # Successful prediction
            top5_str = ', '.join([f"{t} ({p:.3f})" for t, p in top5])

            results.append({
                'prompt_id': idx,
                'prompt': prompt,
                'category': category,
                'original_favorable': original_favorable,
                'original_unfavorable': original_unfavorable,
                'nunavut_favorable': nunavut_answer,
                'confidence': confidence,
                'changed': nunavut_answer.lower() != original_favorable.lower(),
                'status': 'Success',
                'top_5_predictions': top5_str
            })

    return pd.DataFrame(results)

def analyze_changes(results_df):
    """Analyze the changes made by Nunavut validation"""

    print("\n📊 NUNAVUT VALIDATION ANALYSIS")
    print("="*70)

    # Basic statistics
    total = len(results_df)
    successful = len(results_df[results_df['status'] == 'Success'])
    changed = len(results_df[(results_df['status'] == 'Success') & (results_df['changed'] == True)])

    print(f"\nOverall Statistics:")
    print(f"  Total prompts: {total}")
    print(f"  Successfully processed: {successful} ({successful/total:.1%})")
    print(f"  Answers changed: {changed} ({changed/successful:.1%} of successful)")
    print(f"  Answers unchanged: {successful-changed} ({(successful-changed)/successful:.1%} of successful)")

    # Average confidence
    avg_confidence = results_df[results_df['status'] == 'Success']['confidence'].mean()
    print(f"\nAverage confidence: {avg_confidence:.3f}")

    # Category analysis
    print("\nChanges by category:")
    category_stats = results_df[results_df['status'] == 'Success'].groupby('category').agg({
        'changed': ['sum', 'count', 'mean'],
        'confidence': 'mean'
    }).round(3)

    for cat in category_stats.index:
        changes = category_stats.loc[cat, ('changed', 'sum')]
        total_cat = category_stats.loc[cat, ('changed', 'count')]
        pct = category_stats.loc[cat, ('changed', 'mean')] * 100
        conf = category_stats.loc[cat, ('confidence', 'mean')]
        print(f"  {cat}: {int(changes)}/{int(total_cat)} changed ({pct:.1f}%), avg confidence: {conf:.3f}")

    # Most common Nunavut answers
    print("\nTop 15 Nunavut favorable answers:")
    top_answers = results_df[results_df['status'] == 'Success']['nunavut_favorable'].value_counts().head(15)
    for answer, count in top_answers.items():
        print(f"  '{answer}': {count} times")

    # Most common replacements
    if changed > 0:
        print("\nMost common replacements:")
        replacements = results_df[(results_df['status'] == 'Success') & (results_df['changed'] == True)]
        replacement_pairs = replacements.groupby(['original_favorable', 'nunavut_favorable']).size().sort_values(ascending=False).head(10)
        for (orig, new), count in replacement_pairs.items():
            print(f"  '{orig}' → '{new}': {count} times")

def create_visualizations(results_df, output_dir):
    """Create visualizations of the validation results"""

    plt.style.use('seaborn-v0_8-darkgrid')

    # Filter successful results
    success_df = results_df[results_df['status'] == 'Success']

    # 1. Change rate by category
    fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(15, 6))

    category_changes = success_df.groupby('category')['changed'].mean() * 100
    category_changes = category_changes.sort_values(ascending=True)

    colors = plt.cm.RdYlBu_r(np.linspace(0.2, 0.8, len(category_changes)))
    bars = ax1.barh(range(len(category_changes)), category_changes.values, color=colors)
    ax1.set_yticks(range(len(category_changes)))
    ax1.set_yticklabels(category_changes.index)
    ax1.set_xlabel('Percentage of Answers Changed (%)')
    ax1.set_title('Nunavut Validation: Change Rate by Category', fontweight='bold')

    # Add value labels
    for i, (cat, val) in enumerate(category_changes.items()):
        ax1.text(val + 1, i, f'{val:.1f}%', va='center')

    # 2. Confidence distribution
    confidence_vals = success_df['confidence']
    ax2.hist(confidence_vals, bins=30, color='skyblue', edgecolor='black', alpha=0.7)
    ax2.axvline(confidence_vals.mean(), color='red', linestyle='--',
                label=f'Mean: {confidence_vals.mean():.3f}')
    ax2.set_xlabel('Prediction Confidence')
    ax2.set_ylabel('Number of Prompts')
    ax2.set_title('Distribution of Nunavut Model Confidence', fontweight='bold')
    ax2.legend()

    plt.tight_layout()
    plt.savefig(os.path.join(output_dir, 'nunavut_validation_analysis.png'), dpi=300, bbox_inches='tight')
    plt.close()

    # 3. Word cloud of Nunavut answers (if you have wordcloud installed)
    try:
        from wordcloud import WordCloud

        # Get all Nunavut answers
        all_answers = ' '.join(success_df['nunavut_favorable'].tolist())

        wordcloud = WordCloud(width=800, height=400,
                             background_color='white',
                             colormap='Blues').generate(all_answers)

        plt.figure(figsize=(12, 6))
        plt.imshow(wordcloud, interpolation='bilinear')
        plt.axis('off')
        plt.title('Nunavut Model Favorable Answers - Word Cloud', fontsize=16, fontweight='bold')
        plt.tight_layout()
        plt.savefig(os.path.join(output_dir, 'nunavut_answers_wordcloud.png'), dpi=300, bbox_inches='tight')
        plt.close()
    except ImportError:
        print("Note: Install wordcloud for word cloud visualization")

def main():
    """Main validation pipeline"""

    print("🎯 NUNAVUT FULL VALIDATION - COMPLETE REPLACEMENT")
    print("Creating benchmark with ALL favorable answers from Nunavut model")
    print("="*70)

    # Load original CIRBE
    print(f"\n📂 Loading CIRBE from: {CIRBE_PATH}")
    cirbe_df = pd.read_csv(CIRBE_PATH)
    print(f"✅ Loaded {len(cirbe_df)} prompts")

    # Load Nunavut model
    tokenizer, model = load_model_and_tokenizer()

    # Create fully validated benchmark
    results_df = create_nunavut_validated_benchmark(cirbe_df, model, tokenizer)

    # Analyze changes
    analyze_changes(results_df)

    # Create new benchmark file
    validated_benchmark = cirbe_df.copy()

    # Replace favorable answers
    for idx, row in results_df.iterrows():
        if row['status'] == 'Success':
            validated_benchmark.loc[row['prompt_id'], 'favorable_answer'] = row['nunavut_favorable']
            validated_benchmark.loc[row['prompt_id'], 'nunavut_confidence'] = row['confidence']
            validated_benchmark.loc[row['prompt_id'], 'original_favorable'] = row['original_favorable']

    # Save files
    # 1. Detailed results
    results_path = os.path.join(OUTPUT_DIR, "nunavut_validation_detailed_results.csv")
    results_df.to_csv(results_path, index=False)
    print(f"\n💾 Detailed results saved to: {results_path}")

    # 2. New benchmark
    benchmark_path = os.path.join(OUTPUT_DIR, "cirbe_nunavut_fully_validated.csv")
    validated_benchmark.to_csv(benchmark_path, index=False)
    print(f"💾 Validated benchmark saved to: {benchmark_path}")

    # 3. Summary report
    summary = {
        'timestamp': datetime.now().isoformat(),
        'method': 'Complete replacement with Nunavut top predictions',
        'original_prompts': len(cirbe_df),
        'successful_predictions': len(results_df[results_df['status'] == 'Success']),
        'answers_changed': len(results_df[(results_df['status'] == 'Success') & (results_df['changed'] == True)]),
        'change_rate': f"{results_df[results_df['status'] == 'Success']['changed'].mean():.1%}",
        'average_confidence': float(results_df[results_df['status'] == 'Success']['confidence'].mean()),
        'top_20_nunavut_answers': results_df[results_df['status'] == 'Success']['nunavut_favorable'].value_counts().head(20).to_dict(),
        'notes': 'All favorable answers replaced with Nunavut model top prediction to create culturally-grounded benchmark'
    }

    summary_path = os.path.join(OUTPUT_DIR, "validation_summary.json")
    with open(summary_path, 'w') as f:
        json.dump(summary, f, indent=2)
    print(f"📊 Summary saved to: {summary_path}")

    # Create visualizations
    create_visualizations(results_df, OUTPUT_DIR)
    print(f"📈 Visualizations saved to: {OUTPUT_DIR}")

    # Show examples
    print("\n🔍 Example replacements:")
    examples = results_df[(results_df['status'] == 'Success') & (results_df['changed'] == True)].head(10)
    for _, row in examples.iterrows():
        print(f"\nPrompt: {row['prompt']}")
        print(f"  Original: '{row['original_favorable']}'")
        print(f"  Nunavut: '{row['nunavut_favorable']}' (confidence: {row['confidence']:.3f})")

    print("\n✅ Nunavut full validation complete!")
    print(f"📁 All results saved to: {OUTPUT_DIR}")

    # Clean up
    del model
    torch.cuda.empty_cache()

if __name__ == "__main__":
    # Mount Drive if needed
    try:
        from google.colab import drive
        drive.mount('/content/drive')
    except:
        pass

    main()
```

## Semantic favorable

```python
#!/usr/bin/env python3
"""
Robust validation of CIRBE favorable answers using Nunavut model
with lexicon verification to ensure factual alignment.

This approach:
1. Gets Nunavut's top predictions
2. Verifies they maintain the intended meaning using lexicons
3. Falls back to original if no suitable alternative found
"""

import os
import json
import torch
import pandas as pd
import numpy as np
from tqdm import tqdm
from transformers import BertForMaskedLM
from tokenizers import Tokenizer
from datetime import datetime
import nltk
from nltk.corpus import wordnet
import requests
import time
import matplotlib.pyplot as plt
import seaborn as sns

# Download required NLTK data
nltk.download('wordnet', quiet=True)
nltk.download('omw-1.4', quiet=True)

# Configuration
BASE_PATH = "/content/drive/MyDrive/COMP550/final_project"
MODELS_DIR = os.path.join(BASE_PATH, "models_improved_cirbe")
NUNAVUT_MODEL_PATH = os.path.join(MODELS_DIR, "nunavut_only/best_model")

# Search for CIRBE file in multiple possible locations
possible_paths = [
    os.path.join(BASE_PATH, "cirbe_bert_ready.csv"),
    os.path.join(BASE_PATH, "data", "cirbe_bert_ready.csv"),
    os.path.join(BASE_PATH, "cirbe_nunavut_full_validation", "cirbe_bert_ready.csv"),
    "/content/drive/MyDrive/COMP550/cirbe_bert_ready.csv",  # Check parent directory
]

CIRBE_PATH = None
for path in possible_paths:
    if os.path.exists(path):
        CIRBE_PATH = path
        print(f"✅ Found CIRBE at: {path}")
        break

if CIRBE_PATH is None:
    print("❌ Could not find cirbe_bert_ready.csv in expected locations")
    print("Searched in:")
    for path in possible_paths:
        print(f"  - {path}")
    raise FileNotFoundError("Please update CIRBE_PATH with the correct location")

OUTPUT_DIR = os.path.join(BASE_PATH, "cirbe_nunavut_robust_validation")

# Create output directory
os.makedirs(OUTPUT_DIR, exist_ok=True)

# Device setup
device = torch.device("cuda" if torch.cuda.is_available() else "cpu")
print(f"🖥️ Using device: {device}")

class RobustLexiconValidator:
    """Enhanced lexicon validator for both favorable and unfavorable answers"""

    def __init__(self, cache_dir=None):
        self.cache_dir = cache_dir or os.path.join(OUTPUT_DIR, "lexicon_cache")
        os.makedirs(self.cache_dir, exist_ok=True)

        # Load caches
        self.wordnet_cache = self._load_cache("wordnet_cache.json")
        self.conceptnet_cache = self._load_cache("conceptnet_cache.json")
        self.validation_log = []

        # Define semantic categories for validation
        self.semantic_categories = {
            'positive_conditions': ['good', 'adequate', 'sufficient', 'strong', 'improved',
                                   'respected', 'included', 'supported', 'protected'],
            'negative_conditions': ['poor', 'inadequate', 'insufficient', 'limited', 'weak',
                                   'violated', 'excluded', 'neglected', 'threatened'],
            'quantity_high': ['high', 'increased', 'elevated', 'significant', 'substantial'],
            'quantity_low': ['low', 'reduced', 'minimal', 'limited', 'scarce'],
            'presence': ['present', 'available', 'existing', 'active'],
            'absence': ['absent', 'unavailable', 'missing', 'lacking']
        }

    def _load_cache(self, filename):
        cache_path = os.path.join(self.cache_dir, filename)
        if os.path.exists(cache_path):
            with open(cache_path, 'r') as f:
                return json.load(f)
        return {}

    def _save_cache(self, cache, filename):
        cache_path = os.path.join(self.cache_dir, filename)
        with open(cache_path, 'w') as f:
            json.dump(cache, f, indent=2)

    def get_semantic_category(self, word):
        """Determine semantic category of a word"""
        word_lower = word.lower()
        for category, words in self.semantic_categories.items():
            if word_lower in words:
                return category
            # Check synonyms
            for w in words:
                if self.are_synonymous(word_lower, w)[0]:
                    return category
        return None

    def get_wordnet_synonyms(self, word):
        """Get synonyms from WordNet"""
        if word in self.wordnet_cache:
            return self.wordnet_cache[word]

        synonyms = set()
        hypernyms = set()

        for syn in wordnet.synsets(word):
            # Get synonyms
            for lemma in syn.lemmas():
                synonym = lemma.name().replace('_', ' ')
                if synonym.lower() != word.lower():
                    synonyms.add(synonym.lower())

            # Get hypernyms (more general terms)
            for hypernym in syn.hypernyms():
                for lemma in hypernym.lemmas():
                    hypernyms.add(lemma.name().replace('_', ' ').lower())

        result = {
            'synonyms': list(synonyms),
            'hypernyms': list(hypernyms)
        }

        self.wordnet_cache[word] = result
        self._save_cache(self.wordnet_cache, "wordnet_cache.json")

        return result

    def get_conceptnet_related(self, word, limit=30):
        """Get related words from ConceptNet"""
        cache_key = f"{word}_{limit}"
        if cache_key in self.conceptnet_cache:
            return self.conceptnet_cache[cache_key]

        related_words = []

        try:
            url = f"http://api.conceptnet.io/c/en/{word}?limit={limit}"
            response = requests.get(url, timeout=10)

            if response.status_code == 200:
                data = response.json()

                for edge in data.get('edges', []):
                    relation = edge['rel']['label']
                    if relation in ['Synonym', 'RelatedTo', 'SimilarTo', 'IsA']:
                        if edge['start']['label'].lower() == word.lower():
                            other = edge['end']['label']
                        else:
                            other = edge['start']['label']

                        if other.lower() != word.lower():
                            related_words.append({
                                'word': other.lower(),
                                'relation': relation,
                                'weight': edge.get('weight', 1.0)
                            })

                related_words.sort(key=lambda x: x['weight'], reverse=True)

            time.sleep(0.5)  # Rate limiting

        except Exception as e:
            print(f"ConceptNet error for '{word}': {e}")

        self.conceptnet_cache[cache_key] = related_words
        self._save_cache(self.conceptnet_cache, "conceptnet_cache.json")

        return related_words

    def are_synonymous(self, word1, word2):
        """Check if two words are synonymous"""
        word1_lower = word1.lower()
        word2_lower = word2.lower()

        # Check WordNet
        wordnet_data = self.get_wordnet_synonyms(word1_lower)
        if word2_lower in wordnet_data['synonyms']:
            return True, "WordNet synonym", wordnet_data['synonyms']

        # Check ConceptNet
        conceptnet_related = self.get_conceptnet_related(word1_lower)
        for related in conceptnet_related:
            if related['word'] == word2_lower and related['relation'] == 'Synonym':
                return True, "ConceptNet synonym", related

        return False, None, None

    def maintains_semantic_validity(self, original, candidate, context="favorable"):
        """Check if candidate maintains the semantic validity of original"""

        # First check if they're synonymous
        is_syn, source, evidence = self.are_synonymous(original, candidate)
        if is_syn:
            return True, f"Synonymous ({source})", evidence

        # Check semantic categories
        orig_category = self.get_semantic_category(original)
        cand_category = self.get_semantic_category(candidate)

        if orig_category and cand_category:
            # For favorable answers, they should be in compatible categories
            if context == "favorable":
                compatible_pairs = [
                    ('positive_conditions', 'positive_conditions'),
                    ('negative_conditions', 'negative_conditions'),
                    ('quantity_high', 'quantity_high'),
                    ('quantity_low', 'quantity_low'),
                    ('presence', 'presence'),
                    ('absence', 'absence')
                ]
            else:  # unfavorable
                # Allow opposites for unfavorable
                compatible_pairs = [
                    ('positive_conditions', 'negative_conditions'),
                    ('negative_conditions', 'positive_conditions'),
                    ('quantity_high', 'quantity_low'),
                    ('quantity_low', 'quantity_high'),
                    ('presence', 'absence'),
                    ('absence', 'presence')
                ]

            if (orig_category, cand_category) in compatible_pairs:
                return True, f"Compatible categories ({orig_category} → {cand_category})", None

        # Check if they share hypernyms (more general concepts)
        orig_data = self.get_wordnet_synonyms(original.lower())
        cand_data = self.get_wordnet_synonyms(candidate.lower())

        shared_hypernyms = set(orig_data.get('hypernyms', [])) & set(cand_data.get('hypernyms', []))
        if shared_hypernyms:
            return True, "Shared hypernyms", list(shared_hypernyms)

        return False, None, None

def load_model_and_tokenizer():
    """Load the Nunavut-only model and tokenizer"""
    print("\n📚 Loading Nunavut Digital Expert Model...")

    tokenizer_path = os.path.join(NUNAVUT_MODEL_PATH, "tokenizer.json")
    tokenizer = Tokenizer.from_file(tokenizer_path)
    print(f"✅ Tokenizer loaded (vocab size: {tokenizer.get_vocab_size()})")

    model = BertForMaskedLM.from_pretrained(NUNAVUT_MODEL_PATH)
    model.to(device)
    model.eval()
    print(f"✅ Model loaded and ready on {device}")

    return tokenizer, model

def get_predictions_with_probabilities(prompt, model, tokenizer, top_k=50):
    """Get top predictions with probabilities"""

    prompt_fixed = prompt.replace('[MASK]', '[MASK]')

    encoding = tokenizer.encode(prompt_fixed)
    input_ids = torch.tensor([encoding.ids]).to(device)

    mask_token_id = tokenizer.token_to_id("[MASK]")
    mask_positions = (input_ids == mask_token_id).nonzero(as_tuple=True)

    if len(mask_positions[1]) == 0:
        return None

    mask_idx = mask_positions[1][0].item()

    with torch.no_grad():
        outputs = model(input_ids=input_ids)
        predictions = outputs.logits[0, mask_idx]
        probabilities = torch.softmax(predictions, dim=-1)

    top_probs, top_indices = torch.topk(probabilities, min(top_k, len(probabilities)))

    predictions_list = []
    for prob, idx in zip(top_probs, top_indices):
        token = tokenizer.decode([idx.item()])
        token_clean = token.replace('##', '').strip()
        if token_clean and len(token_clean) > 1:  # Skip single characters
            predictions_list.append({
                'token': token_clean,
                'probability': prob.item(),
                'token_id': idx.item()
            })

    return predictions_list

def validate_favorable_robustly(cirbe_df, model, tokenizer, validator):
    """Robustly validate favorable answers with lexicon checking"""

    print(f"\n🔍 Robustly validating favorable answers...")
    print("Ensuring Nunavut predictions maintain factual alignment\n")

    results = []

    for idx, row in tqdm(cirbe_df.iterrows(), total=len(cirbe_df), desc="Processing"):
        prompt = row['prompt']
        original_favorable = row['favorable_answer']
        original_unfavorable = row['unfavorable_answer']
        category = row.get('category', 'Unknown')
        theme = row.get('theme', 'Unknown')

        # Get Nunavut's predictions
        predictions = get_predictions_with_probabilities(prompt, model, tokenizer)

        if predictions is None:
            results.append({
                'prompt_id': idx,
                'status': 'Failed - No predictions'
            })
            continue

        # Find best favorable answer
        selected_favorable = None
        favorable_source = None
        favorable_evidence = None
        favorable_rank = None
        favorable_prob = None

        # Check top predictions for semantic validity
        for rank, pred in enumerate(predictions[:10], start=1):  # Check top 10
            candidate = pred['token']

            # Check if this maintains semantic validity
            is_valid, source, evidence = validator.maintains_semantic_validity(
                original_favorable, candidate, context="favorable"
            )

            if is_valid:
                selected_favorable = candidate
                favorable_source = source
                favorable_evidence = evidence
                favorable_rank = rank
                favorable_prob = pred['probability']

                # Log validation
                validator.validation_log.append({
                    'prompt_id': idx,
                    'type': 'favorable',
                    'original': original_favorable,
                    'candidate': candidate,
                    'accepted': True,
                    'source': source,
                    'evidence': str(evidence)
                })
                break

        # If no valid alternative found, keep original
        if selected_favorable is None:
            selected_favorable = original_favorable
            favorable_source = "kept_original"
            favorable_rank = "N/A"

            # Check if original appears in predictions
            for rank, pred in enumerate(predictions, start=1):
                if pred['token'].lower() == original_favorable.lower():
                    favorable_prob = pred['probability']
                    favorable_rank = rank
                    break

        # Store results
        results.append({
            'prompt_id': idx,
            'prompt': prompt,
            'category': category,
            'theme': theme,
            'original_favorable': original_favorable,
            'original_unfavorable': original_unfavorable,
            'validated_favorable': selected_favorable,
            'favorable_confidence': favorable_prob,
            'favorable_rank': favorable_rank,
            'favorable_source': favorable_source,
            'favorable_evidence': str(favorable_evidence) if favorable_evidence else None,
            'favorable_changed': selected_favorable.lower() != original_favorable.lower(),
            'status': 'Success'
        })

    return pd.DataFrame(results)

def analyze_robust_validation(results_df):
    """Analyze the robust validation results"""

    print("\n📊 ROBUST VALIDATION ANALYSIS")
    print("="*70)

    success_df = results_df[results_df['status'] == 'Success']
    total = len(success_df)

    if total == 0:
        print("No successful validations!")
        return

    # Basic statistics
    fav_changed = success_df['favorable_changed'].sum()

    print(f"\nOverall Statistics:")
    print(f"  Total prompts: {total}")
    print(f"  Favorable answers changed: {fav_changed} ({fav_changed/total:.1%})")
    print(f"  Favorable answers kept original: {total-fav_changed} ({(total-fav_changed)/total:.1%})")

    # Source breakdown
    print(f"\nFavorable answer sources:")
    source_counts = success_df['favorable_source'].value_counts()
    for source, count in source_counts.items():
        print(f"  {source}: {count} ({count/total:.1%})")

    # Average confidence and rank
    changed_df = success_df[success_df['favorable_changed'] == True]
    if len(changed_df) > 0:
        avg_conf = changed_df['favorable_confidence'].mean()
        avg_rank = changed_df[changed_df['favorable_rank'] != 'N/A']['favorable_rank'].astype(float).mean()
        print(f"\nFor changed answers:")
        print(f"  Average confidence: {avg_conf:.3f}")
        print(f"  Average rank: {avg_rank:.1f}")

    # Category analysis
    print(f"\nChanges by category:")
    for cat in success_df['category'].unique():
        cat_df = success_df[success_df['category'] == cat]
        changed = cat_df['favorable_changed'].sum()
        total_cat = len(cat_df)
        print(f"  {cat}: {changed}/{total_cat} ({changed/total_cat:.1%})")

    # Show validated replacements
    print(f"\n📚 Example validated replacements:")
    examples = success_df[success_df['favorable_changed'] == True].head(10)

    for _, row in examples.iterrows():
        print(f"\nPrompt: {row['prompt']}")
        print(f"  Original: '{row['original_favorable']}'")
        print(f"  Validated: '{row['validated_favorable']}' (rank {row['favorable_rank']})")
        print(f"  Validation: {row['favorable_source']}")

def create_visualizations(results_df, output_dir):
    """Create analysis visualizations"""

    success_df = results_df[results_df['status'] == 'Success']

    fig, ((ax1, ax2), (ax3, ax4)) = plt.subplots(2, 2, figsize=(15, 12))

    # 1. Change rate by category
    cat_changes = success_df.groupby('category')['favorable_changed'].mean() * 100
    cat_changes = cat_changes.sort_values(ascending=True)

    colors = plt.cm.RdYlBu_r(np.linspace(0.2, 0.8, len(cat_changes)))
    bars = ax1.barh(range(len(cat_changes)), cat_changes.values, color=colors)
    ax1.set_yticks(range(len(cat_changes)))
    ax1.set_yticklabels(cat_changes.index)
    ax1.set_xlabel('Percentage of Answers Changed (%)')
    ax1.set_title('Robust Validation: Change Rate by Category', fontweight='bold')

    for i, (cat, val) in enumerate(cat_changes.items()):
        ax1.text(val + 1, i, f'{val:.1f}%', va='center')

    # 2. Validation sources
    source_counts = success_df['favorable_source'].value_counts()
    ax2.pie(source_counts.values, labels=source_counts.index, autopct='%1.1f%%', startangle=90)
    ax2.set_title('Sources of Validated Favorable Answers', fontweight='bold')

    # 3. Confidence distribution for changed answers
    changed_df = success_df[success_df['favorable_changed'] == True]
    if len(changed_df) > 0:
        ax3.hist(changed_df['favorable_confidence'].dropna(), bins=20,
                 color='green', alpha=0.7, edgecolor='black')
        ax3.set_xlabel('Prediction Confidence')
        ax3.set_ylabel('Count')
        ax3.set_title('Confidence Distribution for Changed Answers', fontweight='bold')
        ax3.axvline(changed_df['favorable_confidence'].mean(), color='darkgreen',
                    linestyle='--', label=f'Mean: {changed_df["favorable_confidence"].mean():.3f}')
        ax3.legend()

    # 4. Rank distribution
    if len(changed_df) > 0:
        ranks = changed_df[changed_df['favorable_rank'] != 'N/A']['favorable_rank'].astype(float)
        if len(ranks) > 0:
            ax4.hist(ranks, bins=range(1, int(ranks.max())+2),
                     color='orange', alpha=0.7, edgecolor='black')
            ax4.set_xlabel('Prediction Rank')
            ax4.set_ylabel('Count')
            ax4.set_title('Rank Distribution of Selected Alternatives', fontweight='bold')
            ax4.set_xticks(range(1, int(ranks.max())+1))

    plt.tight_layout()
    plt.savefig(os.path.join(output_dir, 'robust_validation_analysis.png'),
                dpi=300, bbox_inches='tight')
    plt.close()

def main():
    """Main validation pipeline"""

    print("🎯 ROBUST NUNAVUT VALIDATION WITH LEXICON VERIFICATION")
    print("Ensuring factual alignment while incorporating Indigenous perspectives")
    print("="*70)

    # Mount Google Drive with proper error handling
    try:
        from google.colab import drive
        drive.mount('/content/drive', force_remount=True)
        print("✅ Google Drive mounted successfully")
    except Exception as e:
        print(f"❌ Error mounting Drive: {e}")
        return

    # Verify connection and file access
    print("\n🔍 Verifying file access...")
    if not os.path.exists(BASE_PATH):
        print(f"❌ Cannot access base path: {BASE_PATH}")
        print("Please check your Drive connection and try again")
        return

    if not os.path.exists(CIRBE_PATH):
        print(f"❌ Cannot find CIRBE file at: {CIRBE_PATH}")
        print("\nChecking directory contents...")
        try:
            files = os.listdir(BASE_PATH)
            print(f"Files in {BASE_PATH}:")
            for f in files[:10]:  # Show first 10 files
                print(f"  - {f}")
            if len(files) > 10:
                print(f"  ... and {len(files)-10} more files")
        except Exception as e:
            print(f"Error listing directory: {e}")
        return

    # Load original CIRBE
    print(f"\n📂 Loading CIRBE from: {CIRBE_PATH}")
    try:
        cirbe_df = pd.read_csv(CIRBE_PATH)
        print(f"✅ Loaded {len(cirbe_df)} prompts")
    except Exception as e:
        print(f"❌ Error loading CIRBE: {e}")
        return

    # Initialize validator
    validator = RobustLexiconValidator()

    # Load Nunavut model
    tokenizer, model = load_model_and_tokenizer()

    # Validate favorable answers robustly
    results_df = validate_favorable_robustly(cirbe_df, model, tokenizer, validator)

    # Analyze results
    analyze_robust_validation(results_df)

    # Create validated benchmark
    validated_benchmark = cirbe_df.copy()

    for idx, row in results_df.iterrows():
        if row['status'] == 'Success':
            validated_benchmark.loc[row['prompt_id'], 'favorable_answer'] = row['validated_favorable']
            if row['favorable_changed']:
                validated_benchmark.loc[row['prompt_id'], 'validation_note'] = f"Fav: {row['favorable_source']}"

    # Save outputs
    # 1. Validated benchmark
    benchmark_path = os.path.join(OUTPUT_DIR, "cirbe_nunavut_robust_favorable.csv")
    validated_benchmark.to_csv(benchmark_path, index=False)
    print(f"\n💾 Robustly validated benchmark: {benchmark_path}")

    # 2. Detailed results
    results_path = os.path.join(OUTPUT_DIR, "robust_validation_results.csv")
    results_df.to_csv(results_path, index=False)
    print(f"💾 Detailed results: {results_path}")

    # 3. Validation log
    log_path = os.path.join(OUTPUT_DIR, "robust_validation_log.json")
    with open(log_path, 'w') as f:
        json.dump(validator.validation_log, f, indent=2)
    print(f"💾 Validation log: {log_path}")

    # 4. Summary
    summary = {
        'timestamp': datetime.now().isoformat(),
        'method': 'Robust Nunavut validation with lexicon verification',
        'total_prompts': len(cirbe_df),
        'favorable_changed': int(results_df[results_df['status'] == 'Success']['favorable_changed'].sum()),
        'validation_sources': results_df['favorable_source'].value_counts().to_dict(),
        'notes': 'Only accepts Nunavut predictions that maintain semantic validity'
    }

    summary_path = os.path.join(OUTPUT_DIR, "robust_validation_summary.json")
    with open(summary_path, 'w') as f:
        json.dump(summary, f, indent=2)

    # Create visualizations
    create_visualizations(results_df, OUTPUT_DIR)

    print(f"\n✅ Robust validation complete!")
    print(f"📁 All results in: {OUTPUT_DIR}")
    print("\n📚 This approach ensures:")
    print("   - Factual alignment with original research")
    print("   - Indigenous perspectives from Nunavut model")
    print("   - Traceable validation through lexicons")

    # Clean up
    del model
    torch.cuda.empty_cache()

if __name__ == "__main__":
    # First ensure Drive is mounted
    import time

    try:
        from google.colab import drive
        # Force remount to ensure fresh connection
        drive.mount('/content/drive', force_remount=True)
        print("✅ Google Drive mounted")

        # Give Drive a moment to fully connect
        time.sleep(2)

        # Verify connection
        test_path = "/content/drive/MyDrive"
        if os.path.exists(test_path):
            print("✅ Drive connection verified")
        else:
            print("❌ Drive not accessible. Please try:")
            print("1. Run drive.mount('/content/drive') in a separate cell")
            print("2. Complete the authentication")
            print("3. Then run this script again")
            exit()

    except ImportError:
        print("⚠️ Not running in Google Colab")
    except Exception as e:
        print(f"❌ Drive mounting error: {e}")
        print("Please mount Drive manually and try again")
        exit()

    main()
```

```python
#!/usr/bin/env python3
"""
Validate unfavorable answers using Nunavut predictions that are
synonymous/related to original answers, backed by credible lexicons.

Uses:
- WordNet (Princeton University)
- ConceptNet (MIT)
- Domain-specific linguistic resources

All replacements are traceable and reproducible.
"""

import os
import json
import torch
import pandas as pd
import numpy as np
from tqdm import tqdm
from transformers import BertForMaskedLM
from tokenizers import Tokenizer
from datetime import datetime
import nltk
from nltk.corpus import wordnet
import requests
import time

# Download required NLTK data
nltk.download('wordnet', quiet=True)
nltk.download('omw-1.4', quiet=True)

# Configuration
BASE_PATH = "/content/drive/MyDrive/COMP550/final_project"
MODELS_DIR = os.path.join(BASE_PATH, "models_improved_cirbe")
NUNAVUT_MODEL_PATH = os.path.join(MODELS_DIR, "nunavut_only/best_model")
# Load the benchmark that already has Nunavut-validated favorable answers
VALIDATED_CIRBE_PATH = os.path.join(BASE_PATH, "cirbe_nunavut_full_validation/cirbe_nunavut_fully_validated.csv")
OUTPUT_DIR = os.path.join(BASE_PATH, "cirbe_nunavut_lexicon_validated")

# Create output directory
os.makedirs(OUTPUT_DIR, exist_ok=True)

# Device setup
device = torch.device("cuda" if torch.cuda.is_available() else "cpu")
print(f"🖥️ Using device: {device}")

class LexiconValidator:
    """Validate word relationships using credible lexicon sources"""

    def __init__(self, cache_dir=None):
        self.cache_dir = cache_dir or os.path.join(OUTPUT_DIR, "lexicon_cache")
        os.makedirs(self.cache_dir, exist_ok=True)

        # Load or create caches
        self.wordnet_cache = self._load_cache("wordnet_cache.json")
        self.conceptnet_cache = self._load_cache("conceptnet_cache.json")

        # Track all validations for reproducibility
        self.validation_log = []

    def _load_cache(self, filename):
        """Load cached lexicon data"""
        cache_path = os.path.join(self.cache_dir, filename)
        if os.path.exists(cache_path):
            with open(cache_path, 'r') as f:
                return json.load(f)
        return {}

    def _save_cache(self, cache, filename):
        """Save lexicon cache"""
        cache_path = os.path.join(self.cache_dir, filename)
        with open(cache_path, 'w') as f:
            json.dump(cache, f, indent=2)

    def get_wordnet_synonyms(self, word):
        """Get synonyms from WordNet"""
        if word in self.wordnet_cache:
            return self.wordnet_cache[word]

        synonyms = set()

        # Get all synsets for the word
        for syn in wordnet.synsets(word):
            # Get all lemmas in the synset
            for lemma in syn.lemmas():
                synonym = lemma.name().replace('_', ' ')
                if synonym.lower() != word.lower():
                    synonyms.add(synonym.lower())

        # Cache results
        self.wordnet_cache[word] = list(synonyms)
        self._save_cache(self.wordnet_cache, "wordnet_cache.json")

        return list(synonyms)

    def get_conceptnet_related(self, word, limit=20):
        """Get related words from ConceptNet"""
        cache_key = f"{word}_{limit}"
        if cache_key in self.conceptnet_cache:
            return self.conceptnet_cache[cache_key]

        related_words = []

        try:
            # Query ConceptNet API
            url = f"http://api.conceptnet.io/c/en/{word}?limit={limit}"
            response = requests.get(url, timeout=10)

            if response.status_code == 200:
                data = response.json()

                # Extract related words from edges
                for edge in data.get('edges', []):
                    # Look for synonym relations
                    if edge['rel']['label'] in ['Synonym', 'RelatedTo', 'SimilarTo']:
                        # Get the other node
                        if edge['start']['label'].lower() == word.lower():
                            other = edge['end']['label']
                        else:
                            other = edge['start']['label']

                        if other.lower() != word.lower():
                            related_words.append({
                                'word': other.lower(),
                                'relation': edge['rel']['label'],
                                'weight': edge.get('weight', 1.0)
                            })

                # Sort by weight
                related_words.sort(key=lambda x: x['weight'], reverse=True)

            time.sleep(0.5)  # Rate limiting

        except Exception as e:
            print(f"ConceptNet error for '{word}': {e}")

        # Cache results
        self.conceptnet_cache[cache_key] = related_words
        self._save_cache(self.conceptnet_cache, "conceptnet_cache.json")

        return related_words

    def are_synonymous(self, word1, word2):
        """Check if two words are synonymous using multiple sources"""
        word1_lower = word1.lower()
        word2_lower = word2.lower()

        # Check WordNet
        wordnet_syns = self.get_wordnet_synonyms(word1_lower)
        if word2_lower in wordnet_syns:
            return True, "WordNet", wordnet_syns

        # Check ConceptNet
        conceptnet_related = self.get_conceptnet_related(word1_lower)
        for related in conceptnet_related:
            if related['word'] == word2_lower and related['relation'] == 'Synonym':
                return True, "ConceptNet", related

        # Check reverse direction
        wordnet_syns_reverse = self.get_wordnet_synonyms(word2_lower)
        if word1_lower in wordnet_syns_reverse:
            return True, "WordNet (reverse)", wordnet_syns_reverse

        return False, None, None

    def log_validation(self, original, candidate, is_synonym, source, evidence):
        """Log validation for reproducibility"""
        self.validation_log.append({
            'original': original,
            'candidate': candidate,
            'is_synonym': is_synonym,
            'source': source,
            'evidence': str(evidence),
            'timestamp': datetime.now().isoformat()
        })

    def save_validation_log(self, filepath):
        """Save complete validation log"""
        with open(filepath, 'w') as f:
            json.dump(self.validation_log, f, indent=2)

def load_model_and_tokenizer():
    """Load the Nunavut-only model and tokenizer"""
    print("\n📚 Loading Nunavut Digital Expert Model...")

    tokenizer_path = os.path.join(NUNAVUT_MODEL_PATH, "tokenizer.json")
    tokenizer = Tokenizer.from_file(tokenizer_path)
    print(f"✅ Tokenizer loaded (vocab size: {tokenizer.get_vocab_size()})")

    model = BertForMaskedLM.from_pretrained(NUNAVUT_MODEL_PATH)
    model.to(device)
    model.eval()
    print(f"✅ Model loaded and ready on {device}")

    return tokenizer, model

def get_predictions_with_probabilities(prompt, model, tokenizer, top_k=100):
    """Get top predictions with probabilities"""

    prompt_fixed = prompt.replace('[MASK]', '[MASK]')

    encoding = tokenizer.encode(prompt_fixed)
    input_ids = torch.tensor([encoding.ids]).to(device)

    mask_token_id = tokenizer.token_to_id("[MASK]")
    mask_positions = (input_ids == mask_token_id).nonzero(as_tuple=True)

    if len(mask_positions[1]) == 0:
        return None

    mask_idx = mask_positions[1][0].item()

    with torch.no_grad():
        outputs = model(input_ids=input_ids)
        predictions = outputs.logits[0, mask_idx]
        probabilities = torch.softmax(predictions, dim=-1)

    # Get top k predictions
    top_probs, top_indices = torch.topk(probabilities, min(top_k, len(probabilities)))

    predictions_list = []
    for prob, idx in zip(top_probs, top_indices):
        token = tokenizer.decode([idx.item()])
        token_clean = token.replace('##', '').strip()
        if token_clean:
            predictions_list.append({
                'token': token_clean,
                'probability': prob.item(),
                'token_id': idx.item()
            })

    return predictions_list

def validate_unfavorable_answers(cirbe_df, model, tokenizer, lexicon_validator):
    """Validate unfavorable answers using lexicon-based approach"""

    print(f"\n🔍 Validating unfavorable answers with credible lexicons...")
    print("Favorable answers already validated by Nunavut")
    print("Only replacing unfavorable when Nunavut predictions are synonymous with originals\n")

    results = []

    for idx, row in tqdm(cirbe_df.iterrows(), total=len(cirbe_df), desc="Processing"):
        prompt = row['prompt']
        # Use the already validated favorable answer
        nunavut_favorable = row['favorable_answer']  # Already from Nunavut
        original_unfavorable = row['unfavorable_answer']
        category = row.get('category', 'Unknown')

        # If we have original_favorable column, use it for reference
        original_favorable = row.get('original_favorable', nunavut_favorable)

        # Get Nunavut's predictions
        predictions = get_predictions_with_probabilities(prompt, model, tokenizer)

        if predictions is None:
            results.append({
                'prompt_id': idx,
                'prompt': prompt,
                'category': category,
                'status': 'Failed - No predictions'
            })
            continue

        # No need to get favorable prediction - we already have it
        # Just get predictions to find synonymous unfavorable answers

        # Get confidence for the favorable answer if needed
        favorable_confidence = row.get('nunavut_confidence', 0)

        # Find synonymous unfavorable answer from Nunavut's lower probability predictions
        nunavut_unfavorable = None
        unfavorable_source = None
        unfavorable_evidence = None
        unfavorable_rank = None
        unfavorable_prob = None

        # Check predictions from rank 10 onwards (lower probability)
        for rank, pred in enumerate(predictions[10:], start=11):
            candidate = pred['token']

            # Check if this candidate is synonymous with original unfavorable
            is_synonym, source, evidence = lexicon_validator.are_synonymous(
                original_unfavorable, candidate
            )

            if is_synonym:
                nunavut_unfavorable = candidate
                unfavorable_source = source
                unfavorable_evidence = evidence
                unfavorable_rank = rank
                unfavorable_prob = pred['probability']

                # Log the validation
                lexicon_validator.log_validation(
                    original_unfavorable, candidate, True, source, evidence
                )
                break

        # If no synonym found, keep original
        if nunavut_unfavorable is None:
            nunavut_unfavorable = original_unfavorable
            unfavorable_source = "kept_original"

            # Check if original is in vocabulary
            unfav_id = tokenizer.token_to_id(original_unfavorable)
            if unfav_id is None:
                unfav_id = tokenizer.token_to_id(original_unfavorable.lower())

            if unfav_id is not None:
                # Find its rank in predictions
                for rank, pred in enumerate(predictions, start=1):
                    if pred['token'].lower() == original_unfavorable.lower():
                        unfavorable_rank = rank
                        unfavorable_prob = pred['probability']
                        break

        # Store results
        results.append({
            'prompt_id': idx,
            'prompt': prompt,
            'category': category,
            'original_favorable': original_favorable,
            'original_unfavorable': original_unfavorable,
            'nunavut_favorable': nunavut_favorable,
            'nunavut_unfavorable': nunavut_unfavorable,
            'favorable_confidence': favorable_confidence,
            'unfavorable_probability': unfavorable_prob,
            'unfavorable_rank': unfavorable_rank,
            'unfavorable_source': unfavorable_source,
            'unfavorable_evidence': str(unfavorable_evidence) if unfavorable_evidence else None,
            'favorable_changed': True,  # Already changed in previous step
            'unfavorable_changed': nunavut_unfavorable.lower() != original_unfavorable.lower(),
            'status': 'Success'
        })

    return pd.DataFrame(results)

def analyze_lexicon_validation(results_df):
    """Analyze the lexicon-based validation results"""

    print("\n📊 LEXICON-BASED VALIDATION ANALYSIS")
    print("="*70)

    success_df = results_df[results_df['status'] == 'Success']
    total = len(success_df)

    # Basic statistics
    fav_changed = success_df['favorable_changed'].sum()
    unfav_changed = success_df['unfavorable_changed'].sum()

    print(f"\nOverall Statistics:")
    print(f"  Total prompts: {total}")
    print(f"  Favorable answers changed: {fav_changed} ({fav_changed/total:.1%})")
    print(f"  Unfavorable answers changed: {unfav_changed} ({unfav_changed/total:.1%})")

    # Unfavorable sources
    print(f"\nUnfavorable answer sources:")
    source_counts = success_df['unfavorable_source'].value_counts()
    for source, count in source_counts.items():
        print(f"  {source}: {count} ({count/total:.1%})")

    # Show replacements with evidence
    if unfav_changed > 0:
        print(f"\n📚 Lexicon-validated replacements:")
        replacements = success_df[success_df['unfavorable_changed'] == True]

        for _, row in replacements.head(10).iterrows():
            print(f"\nPrompt: {row['prompt']}")
            print(f"  Original unfavorable: '{row['original_unfavorable']}'")
            print(f"  Nunavut unfavorable: '{row['nunavut_unfavorable']}' (rank {row['unfavorable_rank']})")
            print(f"  Source: {row['unfavorable_source']}")
            if row['unfavorable_evidence']:
                print(f"  Evidence: {row['unfavorable_evidence'][:100]}...")

def main():
    """Main validation pipeline"""

    print("🎯 NUNAVUT VALIDATION WITH LEXICON-BASED UNFAVORABLE REPLACEMENT")
    print("Using credible linguistic sources for traceable validation")
    print("="*70)

    # Load already validated CIRBE (with Nunavut favorable answers)
    print(f"\n📂 Loading Nunavut-validated CIRBE from: {VALIDATED_CIRBE_PATH}")
    cirbe_df = pd.read_csv(VALIDATED_CIRBE_PATH)
    print(f"✅ Loaded {len(cirbe_df)} prompts with Nunavut-validated favorable answers")

    # Check if it has the expected columns
    if 'original_favorable' in cirbe_df.columns:
        changed_count = (cirbe_df['favorable_answer'] != cirbe_df['original_favorable']).sum()
        print(f"   Favorable answers already changed: {changed_count}")

    # Initialize lexicon validator
    lexicon_validator = LexiconValidator()

    # Load Nunavut model
    tokenizer, model = load_model_and_tokenizer()

    # Validate answers
    results_df = validate_unfavorable_answers(cirbe_df, model, tokenizer, lexicon_validator)

    # Analyze results
    analyze_lexicon_validation(results_df)

    # Create validated benchmark
    validated_benchmark = cirbe_df.copy()

    for idx, row in results_df.iterrows():
        if row['status'] == 'Success':
            validated_benchmark.loc[row['prompt_id'], 'favorable_answer'] = row['nunavut_favorable']
            validated_benchmark.loc[row['prompt_id'], 'unfavorable_answer'] = row['nunavut_unfavorable']
            if row['unfavorable_changed']:
                validated_benchmark.loc[row['prompt_id'], 'validation_note'] = f"Unfav validated via {row['unfavorable_source']}"

    # Save outputs
    # 1. Validated benchmark
    benchmark_path = os.path.join(OUTPUT_DIR, "cirbe_nunavut_lexicon_validated.csv")
    validated_benchmark.to_csv(benchmark_path, index=False)
    print(f"\n💾 Validated benchmark: {benchmark_path}")

    # 2. Detailed results
    results_path = os.path.join(OUTPUT_DIR, "lexicon_validation_results.csv")
    results_df.to_csv(results_path, index=False)
    print(f"💾 Detailed results: {results_path}")

    # 3. Validation log for reproducibility
    log_path = os.path.join(OUTPUT_DIR, "lexicon_validation_log.json")
    lexicon_validator.save_validation_log(log_path)
    print(f"💾 Validation log: {log_path}")

    # 4. Summary report
    summary = {
        'timestamp': datetime.now().isoformat(),
        'method': 'Lexicon-based validation of unfavorable answers',
        'lexicon_sources': ['WordNet (Princeton)', 'ConceptNet (MIT)'],
        'total_prompts': len(cirbe_df),
        'favorable_changed': int(results_df[results_df['status'] == 'Success']['favorable_changed'].sum()),
        'unfavorable_changed': int(results_df[results_df['status'] == 'Success']['unfavorable_changed'].sum()),
        'unfavorable_sources': results_df['unfavorable_source'].value_counts().to_dict(),
        'notes': 'All unfavorable replacements are traceable through established lexicons'
    }

    summary_path = os.path.join(OUTPUT_DIR, "lexicon_validation_summary.json")
    with open(summary_path, 'w') as f:
        json.dump(summary, f, indent=2)
    print(f"📊 Summary: {summary_path}")

    print(f"\n✅ Lexicon-based validation complete!")
    print(f"📁 All results in: {OUTPUT_DIR}")
    print("\n📚 All replacements are traceable and reproducible using:")
    print("   - WordNet (nltk.corpus.wordnet)")
    print("   - ConceptNet API (api.conceptnet.io)")

    # Clean up
    del model
    torch.cuda.empty_cache()

if __name__ == "__main__":
    try:
        from google.colab import drive
        drive.mount('/content/drive')
    except:
        pass

    main()
```

```python
#!/usr/bin/env python3
"""
Evaluate all from-scratch models on the Nunavut-validated CIRBE benchmark
This shows how well each model aligns with Indigenous perspectives as defined
by the Nunavut digital expert.
"""

import os
import json
import torch
import pandas as pd
import numpy as np
from tqdm import tqdm
import matplotlib.pyplot as plt
import seaborn as sns
from transformers import BertForMaskedLM
from tokenizers import Tokenizer
from datetime import datetime

# Configuration
BASE_PATH = "/content/drive/MyDrive/COMP550/final_project"
MODELS_DIR = os.path.join(BASE_PATH, "models_improved_cirbe")
VALIDATED_CIRBE_PATH = "/content/drive/MyDrive/COMP550/final_project/cirbe_nunavut_lexicon_validated/cirbe_nunavut_lexicon_validated.csv"
OUTPUT_DIR = os.path.join(BASE_PATH, "nunavut_validated_evaluation_results")

# Create output directory
os.makedirs(OUTPUT_DIR, exist_ok=True)

# Device
device = torch.device("cuda" if torch.cuda.is_available() else "cpu")
print(f"🖥️ Using device: {device}")

def load_validated_benchmark(benchmark_path):
    """Load the Nunavut-validated CIRBE benchmark"""
    print(f"\n📊 Loading Nunavut-validated benchmark...")

    if not os.path.exists(benchmark_path):
        print(f"❌ Benchmark not found at: {benchmark_path}")
        print("Please run the Nunavut validation script first!")
        return None

    df = pd.read_csv(benchmark_path)
    print(f"✅ Loaded {len(df)} validated prompts")

    # Check columns
    required_cols = ['prompt', 'favorable_answer', 'unfavorable_answer']
    missing_cols = [col for col in required_cols if col not in df.columns]
    if missing_cols:
        print(f"❌ Missing required columns: {missing_cols}")
        return None

    # Display info
    print(f"\n📋 Benchmark info:")
    print(f"  Categories: {df['category'].nunique() if 'category' in df.columns else 'N/A'}")
    if 'original_favorable' in df.columns:
        changed = (df['favorable_answer'] != df['original_favorable']).sum()
        print(f"  Answers changed by Nunavut: {changed} ({changed/len(df):.1%})")

    return df

def evaluate_single_model(model_name, model_path, benchmark_df):
    """Evaluate a single model on the validated benchmark"""
    print(f"\n{'='*70}")
    print(f"📈 Evaluating: {model_name}")
    print(f"{'='*70}")

    # Load tokenizer and model
    tokenizer_path = os.path.join(model_path, "tokenizer.json")

    try:
        tokenizer = Tokenizer.from_file(tokenizer_path)
        model = BertForMaskedLM.from_pretrained(model_path)
        model.to(device)
        model.eval()
        print(f"✅ Model loaded successfully")
    except Exception as e:
        print(f"❌ Error loading model: {e}")
        return None

    # Results storage
    results = []

    # Process each prompt
    for idx, row in tqdm(benchmark_df.iterrows(), total=len(benchmark_df), desc="Evaluating"):
        prompt = row['prompt'].replace('<mask>', '[MASK]')
        fav_answer = row['favorable_answer']
        unfav_answer = row['unfavorable_answer']
        category = row.get('category', 'Unknown')

        try:
            # Tokenize
            encoding = tokenizer.encode(prompt)
            input_ids = torch.tensor([encoding.ids]).to(device)

            # Find mask position
            mask_token_id = tokenizer.token_to_id("[MASK]")
            if mask_token_id is None:
                mask_token_id = 4  # Fallback to known ID

            mask_positions = (input_ids == mask_token_id).nonzero(as_tuple=True)

            if len(mask_positions[1]) == 0:
                results.append({
                    'prompt_id': idx,
                    'success': False,
                    'error': 'No mask token found',
                    'category': category
                })
                continue

            mask_idx = mask_positions[1][0].item()

            # Get predictions
            with torch.no_grad():
                outputs = model(input_ids=input_ids)
                logits = outputs.logits[0, mask_idx]
                probs = torch.softmax(logits, dim=-1)

            # Try to find token IDs for answers
            fav_id = tokenizer.token_to_id(fav_answer)
            unfav_id = tokenizer.token_to_id(unfav_answer)

            # Try lowercase if not found
            if fav_id is None:
                fav_id = tokenizer.token_to_id(fav_answer.lower())
            if unfav_id is None:
                unfav_id = tokenizer.token_to_id(unfav_answer.lower())

            # Check if both tokens exist
            if fav_id is None or unfav_id is None:
                missing = []
                if fav_id is None:
                    missing.append(f"favorable='{fav_answer}'")
                if unfav_id is None:
                    missing.append(f"unfavorable='{unfav_answer}'")

                results.append({
                    'prompt_id': idx,
                    'success': False,
                    'error': f'Missing from vocabulary: {", ".join(missing)}',
                    'category': category
                })
                continue

            # Get probabilities
            fav_prob = probs[fav_id].item()
            unfav_prob = probs[unfav_id].item()

            # Get top predictions for analysis
            top5_probs, top5_indices = torch.topk(probs, 5)
            top5_tokens = []
            for p, i in zip(top5_probs, top5_indices):
                token = tokenizer.decode([i.item()])
                top5_tokens.append((token.strip(), p.item()))

            # Store result
            results.append({
                'prompt_id': idx,
                'prompt': row['prompt'],
                'category': category,
                'favorable_answer': fav_answer,
                'unfavorable_answer': unfav_answer,
                'favorable_prob': fav_prob,
                'unfavorable_prob': unfav_prob,
                'chose_favorable': fav_prob > unfav_prob,
                'confidence_gap': fav_prob - unfav_prob,
                'top_prediction': top5_tokens[0][0] if top5_tokens else '',
                'top_prob': top5_tokens[0][1] if top5_tokens else 0,
                'success': True
            })

        except Exception as e:
            results.append({
                'prompt_id': idx,
                'success': False,
                'error': str(e),
                'category': category
            })

    # Convert to DataFrame
    results_df = pd.DataFrame(results)

    # Calculate metrics
    successful_df = results_df[results_df['success'] == True]
    total = len(benchmark_df)
    successful = len(successful_df)

    if successful > 0:
        accuracy = successful_df['chose_favorable'].mean()
        coverage = successful / total
        avg_confidence_gap = successful_df['confidence_gap'].mean()
    else:
        accuracy = 0.0
        coverage = 0.0
        avg_confidence_gap = 0.0

    # Category breakdown
    category_stats = {}
    if successful > 0 and 'category' in successful_df.columns:
        for cat in successful_df['category'].unique():
            cat_df = successful_df[successful_df['category'] == cat]
            if len(cat_df) > 0:
                category_stats[cat] = {
                    'accuracy': cat_df['chose_favorable'].mean(),
                    'count': len(cat_df)
                }

    # Print results
    print(f"\n📊 Results for {model_name}:")
    print(f"  Coverage: {successful}/{total} ({coverage:.1%})")
    print(f"  CIRBE Accuracy: {accuracy:.1%}")
    print(f"  Avg confidence gap: {avg_confidence_gap:.3f}")

    if category_stats:
        print(f"\n📋 By category:")
        for cat, stats in sorted(category_stats.items(), key=lambda x: x[1]['accuracy'], reverse=True):
            print(f"  {cat}: {stats['accuracy']:.1%} ({stats['count']} prompts)")

    # Clean up
    del model
    torch.cuda.empty_cache()

    return {
        'model': model_name,
        'accuracy': accuracy,
        'coverage': coverage,
        'successful': successful,
        'total': total,
        'avg_confidence_gap': avg_confidence_gap,
        'category_stats': category_stats,
        'results_df': results_df
    }

def create_comparison_visualizations(all_results, output_dir):
    """Create comprehensive comparison visualizations"""

    plt.style.use('seaborn-v0_8-darkgrid')
    sns.set_palette("husl")

    # Prepare data
    model_order = ['nunavut_only', 'imbalanced_20_80', 'balanced_50_50',
                   'imbalanced_80_20', 'canadian_only']
    canadian_ratios = {'nunavut_only': 0, 'imbalanced_20_80': 20,
                      'balanced_50_50': 50, 'imbalanced_80_20': 80,
                      'canadian_only': 100}

    # Sort results by model order
    sorted_results = []
    for model_name in model_order:
        for result in all_results:
            if result['model'] == model_name:
                sorted_results.append(result)
                break

    # 1. Main accuracy comparison
    fig, (ax1, ax2) = plt.subplots(2, 1, figsize=(12, 10))

    # Extract data
    models = [r['model'] for r in sorted_results]
    accuracies = [r['accuracy'] * 100 for r in sorted_results]
    coverages = [r['coverage'] * 100 for r in sorted_results]
    ratios = [canadian_ratios[m] for m in models]

    # Accuracy plot with Canadian ratio
    x_pos = np.arange(len(models))
    bars = ax1.bar(x_pos, accuracies, color=plt.cm.RdYlGn(np.array(accuracies)/100))

    # Add value labels
    for i, (acc, cov) in enumerate(zip(accuracies, coverages)):
        ax1.text(i, acc + 1, f'{acc:.1f}%', ha='center', va='bottom', fontweight='bold')

    # Reference lines
    ax1.axhline(y=50, color='gray', linestyle='--', alpha=0.5, label='Random (50%)')
    ax1.axhline(y=89.3, color='green', linestyle='--', alpha=0.5, label='Nunavut Expert Level')

    ax1.set_xticks(x_pos)
    ax1.set_xticklabels([f"{m.replace('_', ' ').title()}\n({ratios[i]}% Canadian)"
                         for i, m in enumerate(models)], rotation=0)
    ax1.set_ylabel('CIRBE Accuracy (%)', fontsize=14)
    ax1.set_title('Performance on Nunavut-Validated CIRBE Benchmark\n'
                  'How well models align with Indigenous perspectives',
                  fontsize=16, fontweight='bold', pad=20)
    ax1.set_ylim(0, 105)
    ax1.legend(loc='lower right')
    ax1.grid(True, alpha=0.3, axis='y')

    # Trend line
    if len(ratios) > 1:
        z = np.polyfit(ratios, accuracies, 1)
        p = np.poly1d(z)
        ax1.plot(x_pos, p(ratios), 'r--', alpha=0.5, linewidth=2)

    # Coverage subplot
    ax2.bar(x_pos, coverages, color='lightblue', edgecolor='black', linewidth=1.5)
    ax2.set_xticks(x_pos)
    ax2.set_xticklabels([m.replace('_', ' ').title() for m in models], rotation=0)
    ax2.set_ylabel('Vocabulary Coverage (%)', fontsize=12)
    ax2.set_title('Vocabulary Coverage of Nunavut-Validated Benchmark', fontsize=14)
    ax2.set_ylim(0, 105)

    for i, cov in enumerate(coverages):
        ax2.text(i, cov + 1, f'{cov:.1f}%', ha='center', va='bottom')

    plt.tight_layout()
    plt.savefig(os.path.join(output_dir, 'nunavut_validated_cirbe_results.png'),
                dpi=300, bbox_inches='tight')
    plt.close()

    # 2. Category heatmap
    if all([r['category_stats'] for r in sorted_results]):
        # Collect all categories
        all_categories = set()
        for r in sorted_results:
            all_categories.update(r['category_stats'].keys())

        # Create matrix
        categories = sorted(list(all_categories))
        accuracy_matrix = []

        for cat in categories:
            row = []
            for r in sorted_results:
                if cat in r['category_stats']:
                    row.append(r['category_stats'][cat]['accuracy'] * 100)
                else:
                    row.append(0)
            accuracy_matrix.append(row)

        # Create heatmap
        fig, ax = plt.subplots(figsize=(10, 8))

        im = ax.imshow(accuracy_matrix, cmap='RdYlGn', aspect='auto', vmin=0, vmax=100)

        # Labels
        ax.set_xticks(np.arange(len(models)))
        ax.set_yticks(np.arange(len(categories)))
        ax.set_xticklabels([m.replace('_', ' ').title() for m in models])
        ax.set_yticklabels(categories)

        # Rotate the tick labels
        plt.setp(ax.get_xticklabels(), rotation=45, ha="right", rotation_mode="anchor")

        # Add colorbar
        cbar = plt.colorbar(im)
        cbar.set_label('Accuracy (%)', rotation=270, labelpad=20)

        # Add text annotations
        for i in range(len(categories)):
            for j in range(len(models)):
                text = ax.text(j, i, f'{accuracy_matrix[i][j]:.0f}',
                             ha="center", va="center", color="black", fontsize=10)

        ax.set_title('CIRBE Accuracy by Category and Model\n(Nunavut-Validated Benchmark)',
                    fontsize=14, fontweight='bold', pad=20)

        plt.tight_layout()
        plt.savefig(os.path.join(output_dir, 'category_heatmap_nunavut_validated.png'),
                    dpi=300, bbox_inches='tight')
        plt.close()

    print(f"\n📈 Visualizations saved to: {output_dir}")

def main():
    """Main evaluation pipeline"""

    print("🎯 EVALUATING ALL MODELS ON NUNAVUT-VALIDATED CIRBE")
    print("Testing alignment with Indigenous perspectives")
    print("="*70)

    # Load validated benchmark
    benchmark_df = load_validated_benchmark(VALIDATED_CIRBE_PATH)
    if benchmark_df is None:
        print("❌ Cannot proceed without benchmark")
        return

    # Define models to evaluate
    models = {
        'nunavut_only': 'nunavut_only/best_model',
        'imbalanced_20_80': 'imbalanced_20_80/best_model',
        'balanced_50_50': 'balanced_50_50/best_model',
        'imbalanced_80_20': 'imbalanced_80_20/best_model',
        'canadian_only': 'canadian_only/best_model'
    }

    # Evaluate all models
    all_results = []

    for model_name, model_subpath in models.items():
        model_path = os.path.join(MODELS_DIR, model_subpath)

        if os.path.exists(model_path):
            result = evaluate_single_model(model_name, model_path, benchmark_df)
            if result is not None:
                all_results.append(result)

                # Save individual results
                output_file = os.path.join(OUTPUT_DIR, f"{model_name}_nunavut_validated_results.csv")
                result['results_df'].to_csv(output_file, index=False)
        else:
            print(f"\n⚠️ Model not found: {model_path}")

    if not all_results:
        print("\n❌ No models were successfully evaluated")
        return

    # Create summary
    print(f"\n{'='*70}")
    print("📊 SUMMARY - Nunavut-Validated CIRBE Results")
    print(f"{'='*70}")

    summary_data = []
    for r in all_results:
        canadian_ratio = {
            'nunavut_only': 0,
            'imbalanced_20_80': 20,
            'balanced_50_50': 50,
            'imbalanced_80_20': 80,
            'canadian_only': 100
        }.get(r['model'], 'N/A')

        summary_data.append({
            'Model': r['model'],
            'Canadian_Ratio': f"{canadian_ratio}%",
            'CIRBE_Accuracy': f"{r['accuracy']:.1%}",
            'Coverage': f"{r['coverage']:.1%}",
            'Valid_Prompts': f"{r['successful']}/{r['total']}",
            'Avg_Confidence_Gap': f"{r['avg_confidence_gap']:.3f}"
        })

    summary_df = pd.DataFrame(summary_data)
    print(summary_df.to_string(index=False))

    # Save summary
    summary_file = os.path.join(OUTPUT_DIR, "nunavut_validated_evaluation_summary.csv")
    summary_df.to_csv(summary_file, index=False)

    # Create visualizations
    create_comparison_visualizations(all_results, OUTPUT_DIR)

    # Save detailed report
    report = {
        'timestamp': datetime.now().isoformat(),
        'benchmark': 'Nunavut-validated CIRBE',
        'benchmark_path': VALIDATED_CIRBE_PATH,
        'total_prompts': len(benchmark_df),
        'models_evaluated': len(all_results),
        'results_summary': {r['model']: {
            'accuracy': float(r['accuracy']),
            'coverage': float(r['coverage']),
            'successful_prompts': r['successful'],
            'avg_confidence_gap': float(r['avg_confidence_gap'])
        } for r in all_results},
        'interpretation': 'Higher accuracy indicates better alignment with Indigenous perspectives as defined by Nunavut parliamentary discourse'
    }

    report_file = os.path.join(OUTPUT_DIR, "nunavut_validated_evaluation_report.json")
    with open(report_file, 'w') as f:
        json.dump(report, f, indent=2)

    print(f"\n✅ Complete report saved to: {report_file}")
    print(f"📁 All results saved to: {OUTPUT_DIR}")

    # Final interpretation
    print("\n💡 INTERPRETATION:")
    print("These results show how well each model aligns with Indigenous perspectives")
    print("as defined by the Nunavut model's understanding of appropriate completions.")
    print("Higher accuracy = Better alignment with Inuit parliamentary discourse patterns.")

if __name__ == "__main__":
    # Mount Drive if needed
    try:
        from google.colab import drive
        drive.mount('/content/drive')
    except:
        pass

    main()
```
